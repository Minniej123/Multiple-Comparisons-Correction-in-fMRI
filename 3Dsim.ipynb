```python

#We compare a series of metrics, including DICE coefficients, blob size, number of sigificant pixels, convergence rates along with Power FDR and FWER.
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
from scipy import ndimage, stats


#The set up:
#We now try 3D
#setting up our simulation
#reproducability
rng_master = np.random.default_rng(0)

#grid
v = 100 # number of voxels in each dimension, we've reduced from 200 to 100 since now 3D
Nx, Ny,Nz = v, v,v# 100x100x100 grid
Z, Y, X = np.mgrid[0:Nz, 0:Ny, 0:Nx]

alpha = 0.05 #signif level stays the same throughout our simulations
q_fdr = 0.05 #We use this in our FDR approaches, stays the same throughout our simulations
blob_radius= 8 #made smaller from 12 because of smaller grid

n_runs = 50 #Monte Carlo , use 500 in final
p0_cluster = 0.01 #one-sided cluster-forming threshold
n_perm_cluster = 15 #permutations per run, use 1000 for final

sim_struct = np.ones((3, 3,3), dtype=int) #the 3D structure
analysis_mask = np.ones((Nz, Ny, Nx), dtype=bool) #we use ones but don't consider edge cases, essentially 'where our sims run'


#the methods we will compare as discussed in the report:
methods = ["Bonf", "BH", "BY", "BKY", "ClusterPerm"]

#smoothness
signal_fwhm = 8 # we keep signal field smoothness constant, we choose this based on nichols

#we define this as a function as later fwhm changes with scenario
def fwhm_to_tau(fwhm: float) -> float:
    return float(fwhm) / np.sqrt(8.0 * np.log(2.0))

#sd of gaussian signal kernel
sigma_S = fwhm_to_tau(signal_fwhm) #sigma_s parametrised by fwhm=8


#this is how we define our truth/ not truth mask
#def disk_mask(x0, y0, radius):
 #   return ((X - x0)**2 + (Y - y0)**2) <= radius**2

#using the same idea as the disk mask but this time a sphere
def sphere_mask(x0, y0, z0, radius):
    return ((X - x0)**2 + (Y - y0)**2+ (Z-z0)**2) <= radius**2

#we define a function to estimate how much Gaussian smoothing with tau reduces the sd of white noise.
def estimate_shrinkage_ratio(tau, n_rep=200, seed=0):
    rng = np.random.default_rng(seed)
    ratios = []
    for _ in range(n_rep): #we repeatedly generate white noise fields
        white = rng.normal(0.0, 1.0, size=(Nz, Ny, Nx))
        sd0 = white.std(ddof=1)
        sm = ndimage.gaussian_filter(white, sigma=tau, mode="reflect", cval=0.0) if tau > 0 else white
        sd1 = sm.std(ddof=1)
        ratios.append(sd1 / sd0) #and compute ratio of post/pre smooth sd
    return float(np.mean(ratios))

post_smooth_sd_target = 1.0 #what we want the final noise SD to be after smoothing

_shrink_cache = {} # this avoids recomputing every time

#here calc pre noise sd needed
def sigmaN_for_fwhm(fwhm, target_post_sd=post_smooth_sd_target):
    tau = fwhm_to_tau(fwhm)
    if tau == 0:
        return target_post_sd
    key = round(float(tau), 6)
    if key not in _shrink_cache:
        _shrink_cache[key] = estimate_shrinkage_ratio(tau, n_rep=500, seed=0)
    ratio = _shrink_cache[key]
    return target_post_sd / ratio

#we want the blobs to be equally spaced and centred around the center of the grid
def get_blob_centers(n_blobs, radius=20): #made radius smaller from 60 since smaller grid
    cx, cy, cz = Nx / 2.0, Ny / 2.0, Nz / 2.0 #added extra dimension
    if n_blobs == 1: #1 blob in center
        return [(cx, cy, cz)]
    angles = np.linspace(0, 2 * np.pi, n_blobs, endpoint=False)
    return [(cx + radius * np.cos(a),cy + radius * np.sin(a),cz) for a in angles] #3D version

#to compare the effects of merged signals, we create a series of scenarios where we change the blob distance
def centers_merged_three(center_sep=5): # went from 10 to 5 due to smaller grid
    cx, cy, cz = Nx / 2.0, Ny / 2.0, Nz / 2.0
    h = center_sep * np.sqrt(3) / 2
    return [
        (cx,               cy - 2*h/3, cz),
        (cx - center_sep/2, cy + h/3,  cz),
        (cx + center_sep/2, cy + h/3,  cz),
    ]


#Metrics and comparison

def bootstrap_ci_mean(x, n_boot=2000, ci=95, seed=0): #for thorough comparison we implement CI, we use bootstrap because of MC
    x = np.asarray(x, dtype=float)
    x = x[np.isfinite(x)]
    if x.size == 0:
        return np.nan, np.nan, np.nan
    rng = np.random.default_rng(seed)
    boots = rng.choice(x, size=(n_boot, x.size), replace=True).mean(axis=1)
    lo = np.percentile(boots, (100 - ci) / 2)
    hi = np.percentile(boots, 100 - (100 - ci) / 2)
    return float(x.mean()), float(lo), float(hi)

#we add binomial confidence interval and error bars
def binomial_exact_ci(k, n, alpha=0.05):
    k = int(k)
    n = int(n)
    if n <= 0:
        return np.nan, np.nan, np.nan
    p = k / n
    if k == 0:
        lo = 0.0
    else:
        lo = float(stats.beta.ppf(alpha / 2, k, n-k + 1))
    if k == n:
        hi = 1.0
    else:
        hi = float(stats.beta.ppf(1 - alpha / 2, k + 1, n - k))
    return float(p), float(lo), float(hi)

#for running means :
def add_running_stats(df, value_col, group_cols):
    df = df.sort_values(list(group_cols) + ["run"]).copy()
    g = df.groupby(group_cols, sort=False)[value_col]
    idx_levels = list(range(len(group_cols)))

    df[f"{value_col}_running_mean"] = (
        g.expanding().mean().reset_index(level=idx_levels, drop=True)
    )
    running_std = (
        g.expanding().std(ddof=1).reset_index(level=idx_levels, drop=True)
    ).fillna(0.0)

    running_n = g.expanding().count().reset_index(level=idx_levels, drop=True)
    df[f"{value_col}_running_se"] = running_std / np.sqrt(np.maximum(running_n, 1))
    return df

#DICE is a measure of spatial accuracy
def dice(mask, truth):
    a = int(np.sum(mask))
    b = int(np.sum(truth))
    if a == 0 and b == 0:
        return 1.0
    inter = int(np.sum(mask & truth))
    return 2.0 * inter / max(a + b, 1)

#we define a function for
def per_blob_detection_fraction(sig_mask, blob_truth_masks, frac=0.10, min_vox=20):
    if len(blob_truth_masks) == 0:
        return np.nan, np.nan, [] #no blob, no detect
    sig = sig_mask.astype(bool) & analysis_mask
    detected = [] #1 or 0
    for tm in blob_truth_masks:
        tm = tm.astype(bool) & analysis_mask
        denom = int(np.sum(tm))
        if denom == 0:
            detected.append(0)
            continue
        overlap = int(np.sum(sig & tm)) #how many overlapped
        #gives threshold of how many must've been detected
        thresh = int(np.ceil(frac * denom))
        thresh = max(min_vox, thresh)
        thresh = min(denom, thresh)
        detected.append(1 if overlap >= thresh else 0)
    detected_arr = np.asarray(detected, dtype=int)
    return float(detected_arr.mean()), float(1.0 if np.all(detected_arr == 1) else 0.0), detected

#one sided p-val, since only considering areas of activation
def compute_t_p(images, eps=1e-8):
    images = np.asarray(images)
    n = images.shape[0]
    df = n - 1 #degrees of freedom
    mean = images.mean(axis=0)
    sd = images.std(axis=0, ddof=1)
    tmap = mean / np.maximum(sd / np.sqrt(n), eps)
    pmap = stats.t.sf(tmap, df=df)
    pmap = np.clip(pmap, 1e-300, 1.0)
    return tmap, pmap


#permutation test

#cluster perm, sign flip- we use this instead of freedman lane
_SIGNS = np.array([-1.0, 1.0], dtype=float) #sign flip

def cluster_perm_onesample_one_sided(images, t0, alpha, n_perm, mask, rng,
                                     eps=1e-12, chunk=64):
    images = np.asarray(images, dtype=float)
    n, Nz_, Ny_, Nx_ = images.shape #added Nz
    df = n - 1
    t0 = float(t0)
    V = Nz_ * Ny_ * Nx_ #flatten once
    imgs2 = images.reshape(n, V)
    SS = np.sum(imgs2 ** 2, axis=0)#precomputing SS
    # Observed t-map
    mean_obs = imgs2.mean(axis=0)
    var_obs = (SS - n * mean_obs**2) / max(df, 1)
    var_obs = np.maximum(var_obs, 0.0)
    denom_obs = np.sqrt(var_obs / n) + eps
    t_obs = (mean_obs / denom_obs).reshape(Nz_, Ny_, Nx_)

    #null dist
    null_max = np.empty(n_perm, dtype=int)
    signs_all = rng.choice(_SIGNS, size=(n_perm, n), replace=True)
    mask = mask.astype(bool)

    for start in range(0, n_perm, chunk):
        stop = min(n_perm, start + chunk)
        signs = signs_all[start:stop] # we vectorise (B,v)
        mean_perm = (signs @ imgs2) / n  #permuted mean under sign flips
        var_perm = (SS[None, :] - n * mean_perm**2) / max(df, 1) #variance
        var_perm = np.maximum(var_perm, 0.0)
        denom_perm = np.sqrt(var_perm / n) + eps
        t_chunk = (mean_perm / denom_perm).reshape(stop - start, Nz_, Ny_, Nx_)

        # Label clusters for each permuted map
        for b in range(stop - start):
            supra = (t_chunk[b] >= t0) & mask
            lab, nlab = ndimage.label(supra, structure=sim_struct)
            if nlab == 0:
                null_max[start + b] = 0
            else:
                counts = np.bincount(lab.ravel())
                null_max[start + b] = int(counts[1:].max(initial=0))
    supra_obs = (t_obs >= t0) & mask #cluster observed
    lab, nlab = ndimage.label(supra_obs, structure=sim_struct)
    sig = np.zeros_like(supra_obs, dtype=bool)

    if nlab == 0:
        return sig, t0

    sizes_obs = np.bincount(lab.ravel())[1:]  #labelling
    ge = (null_max[:, None] >= sizes_obs[None, :]) #vectorised perm p-vals for each observed cluster size
    pvals = (1.0 + ge.sum(axis=0)) / (n_perm + 1.0)

    # Mark significant clusters
    for k, p in enumerate(pvals, start=1):
        if p <= alpha:
            sig[lab == k] = True
    return sig, t0

#we use blob sizes as a metric for comparison
def blob_size_stats(sig_mask, structure=sim_struct):
    sig = sig_mask.astype(bool) & analysis_mask
    lab, nlab = ndimage.label(sig, structure=structure)
    if nlab == 0:
        return 0, np.nan, np.nan, 0, np.array([], dtype=int)

    sizes = ndimage.sum(np.ones_like(sig, dtype=int), labels=lab, index=np.arange(1, nlab + 1)).astype(int)
    return int(nlab), float(np.mean(sizes)), float(np.median(sizes)), int(np.max(sizes)), sizes


def cluster_level_counts_against_support(mask_sig, truth_mask):
    sig = mask_sig.astype(bool) & analysis_mask
    truth = truth_mask.astype(bool) & analysis_mask

    lab, nlab = ndimage.label(sig, structure=sim_struct)
    R_cluster = int(nlab)
    if nlab == 0:
        return R_cluster, 0
    V_cluster_no_overlap = 0
    for k in range(1, nlab + 1):
        cl = (lab == k)
        if not np.any(cl & truth):
            V_cluster_no_overlap += 1
    return R_cluster, int(V_cluster_no_overlap)




def per_blob_cluster_detection(sig_mask, blob_truth_masks, frac=0.10, min_vox=20, structure=sim_struct):
    sig = sig_mask.astype(bool) & analysis_mask
    lab, nlab = ndimage.label(sig, structure=structure)
    if nlab == 0:
        return np.zeros(len(blob_truth_masks), dtype=int)

    detected = np.zeros(len(blob_truth_masks), dtype=int)
    for j, tm in enumerate(blob_truth_masks):
        tm = tm.astype(bool) & analysis_mask
        denom = int(tm.sum())
        if denom == 0:
            continue
        thresh = int(np.ceil(frac * denom))
        thresh = max(min_vox, min(thresh, denom))

        for k in range(1, nlab + 1):
            cl = (lab == k)
            if int((cl & tm).sum()) >= thresh:
                detected[j] = 1
                break

    return detected



#we create one evaluation script to run all of  our metrics at once
def evaluate_method(mask_sig, truth_mask, true_null_mask, report_fdr=True, report_cluster=False, truth_n=None):
    sig = mask_sig.astype(bool) & analysis_mask
    truth = truth_mask.astype(bool) & analysis_mask
    true_null = true_null_mask.astype(bool) & analysis_mask
    TP = int(np.sum(sig & truth)) #true pos
    FN = int(np.sum((~sig) & truth)) #false neg
    R  = int(np.sum(sig)) #

    power = (TP / (TP + FN)) if (TP + FN) > 0 else np.nan
    precision = (TP / R) if R > 0 else np.nan
    dice_val = float(dice(sig, truth))

    #False pos
    V_true = int(np.sum(sig & true_null))
    FDP_cond = (V_true / R) if R > 0 else np.nan
    FDP_uncond = (V_true / max(R, 1))
    FWER_true_null = float(1.0 if V_true > 0 else 0.0)

    if truth_n is None:
        truth_n = int(np.sum(truth))
    extent_inflation = (float(R) / float(truth_n)) if truth_n > 0 else np.nan

    n_blobs, blob_mean, blob_median, blob_max, _sizes = blob_size_stats(sig, structure=sim_struct)

    out = {
        "R": R,
        "R_gt0": int(R > 0),

        "TP": TP,
        "FN": FN,
        "truth_n": int(truth_n),

        "Power": float(power) if np.isfinite(power) else np.nan,
        "Precision": float(precision) if np.isfinite(precision) else np.nan,
        "Dice": float(dice_val) if np.isfinite(dice_val) else np.nan,

        "V_true": V_true,
        "FDP": float(FDP_uncond) if report_fdr else np.nan,
        "FDP_cond_Rgt0": float(FDP_cond) if report_fdr else np.nan,
        "FWER_voxel_true_null": FWER_true_null,


        "extent_inflation": float(extent_inflation) if np.isfinite(extent_inflation) else np.nan,
        "n_sig_blobs": int(n_blobs),
        "sig_blob_mean_size": float(blob_mean) if np.isfinite(blob_mean) else np.nan,
        "sig_blob_median_size": float(blob_median) if np.isfinite(blob_median) else np.nan,
        "sig_blob_max_size": int(blob_max),
        "R_cluster": np.nan,
        "V_cluster_no_overlap": np.nan,
        "prop_clusters_no_overlap": np.nan,

        #any event
        "ClusterAnySig": np.nan,
        "ClusterFP_any": np.nan,
    }

    if report_cluster:

        # Rcl= num sig clusters, Vcl = num sig clusters with no overlap with truth
        Rcl, Vcl = cluster_level_counts_against_support(mask_sig, truth_mask)

        out["R_cluster"] = int(Rcl)
        out["V_cluster_no_overlap"] = int(Vcl)
        out["prop_clusters_no_overlap"] = float(Vcl / Rcl) if Rcl > 0 else np.nan
        out["ClusterAnySig"] = float(Rcl > 0)
        out["ClusterFP_any"] = float(Vcl > 0)

    return out


#we define helper function for simulation
def simulate_images(S, n_subj, sigma_N, tau, rng):
    imgs = np.empty((n_subj, Nz, Ny, Nx), dtype=float)

    for i in range(n_subj):
        white = rng.normal(0.0, 1.0, size=(Nz, Ny, Nx)) * float(sigma_N)
        if tau > 0:
            white = ndimage.gaussian_filter(white, sigma=tau, mode="reflect", cval=0.0)
        imgs[i] = S + white

    return imgs


#we run and here define our correction methods
def run_one(images, truth_mask, true_null_mask, rng_run, truth_n, blob_truth_masks, alpha, q_fdr, n_perm_cluster, t0_cluster):
    tmap, pmap = compute_t_p(images)
    #define these once here as BH,BKY and BY all use
    p_vec = pmap[analysis_mask].ravel()
    p_vec = p_vec[np.isfinite(p_vec)]
    p_sorted = np.sort(p_vec)
    m = int(p_sorted.size)
    metrics = {}
#These methods are applicable to 3D setting as they use analysis mask and sim mask that has now been defined for 3D.
#using the definitions given in report, we can code these methods up:
    # Bonferroni
    bonf = np.zeros_like(pmap, dtype=bool)
    bonf[analysis_mask] = pmap[analysis_mask] <= (alpha / m)
    metrics["Bonf"] = evaluate_method(
    bonf, truth_mask, true_null_mask,
    report_fdr=False, report_cluster=False,
    truth_n=truth_n)
    #BH
    bh_line = (np.arange(1, m + 1) / m) * q_fdr
    bh_ok = p_sorted <= bh_line
    pcrit_bh = float(p_sorted[np.max(np.where(bh_ok)[0])]) if np.any(bh_ok) else np.nan
    bh = np.zeros_like(pmap, dtype=bool)
    if np.isfinite(pcrit_bh):
        bh[analysis_mask] = pmap[analysis_mask] <= pcrit_bh

    metrics["BH"] = evaluate_method(
    bh, truth_mask, true_null_mask,
    report_fdr=True, report_cluster=False,
    truth_n=truth_n)
    metrics["BH"]["pcrit"] = float(pcrit_bh)

    #BY
    c_m = np.sum(1.0 / np.arange(1, m + 1))
    by_line = (np.arange(1, m + 1) / m) * (q_fdr / c_m)
    by_ok = p_sorted <= by_line
    pcrit_by = float(p_sorted[np.max(np.where(by_ok)[0])]) if np.any(by_ok) else np.nan
    by = np.zeros_like(pmap, dtype=bool)
    if np.isfinite(pcrit_by):
        by[analysis_mask] = pmap[analysis_mask] <= pcrit_by

    metrics["BY"] = evaluate_method(
    by, truth_mask, true_null_mask,
    report_fdr=True, report_cluster=False,
    truth_n=truth_n)
    metrics["BY"]["pcrit"] = float(pcrit_by)

    #BKY
    q0 = q_fdr / (1.0 + q_fdr)
    bky1_line = (np.arange(1, m + 1) / m) * q0
    bky1_ok = p_sorted <= bky1_line
    pcrit1 = float(p_sorted[np.max(np.where(bky1_ok)[0])]) if np.any(bky1_ok) else np.nan
    R1 = int(np.sum(p_vec <= pcrit1)) if np.isfinite(pcrit1) else 0
    m0_hat = max(1, m - R1)
    q_star = min(1.0, q0 * (m / m0_hat))
    #second stage
    bky2_line = (np.arange(1, m + 1) / m) * q_star
    bky2_ok = p_sorted <= bky2_line
    pcrit_bky = float(p_sorted[np.max(np.where(bky2_ok)[0])]) if np.any(bky2_ok) else np.nan

    bky = np.zeros_like(pmap, dtype=bool)
    if np.isfinite(pcrit_bky):
        bky[analysis_mask] = pmap[analysis_mask] <= pcrit_bky

    metrics["BKY"] = evaluate_method(
    bky, truth_mask, true_null_mask,
    report_fdr=True, report_cluster=False,
    truth_n=truth_n)
    metrics["BKY"]["pcrit"] = float(pcrit_bky)

    #Cluster permutation
    seed_perm = int(rng_run.integers(0, 2**32 - 1))
    rng_perm = np.random.default_rng(seed_perm)

    t_start = time.perf_counter()
    clus, t0 = cluster_perm_onesample_one_sided(
        images,
        t0=t0_cluster,
        alpha=alpha,
        n_perm=n_perm_cluster,
        mask=analysis_mask,
        rng=rng_perm,
        chunk=64
    )
    t_elapsed = time.perf_counter() - t_start



    metrics["ClusterPerm"] = evaluate_method(
        clus, truth_mask, true_null_mask,
        report_fdr=False, report_cluster=True,
        truth_n=truth_n
    )
    metrics["ClusterPerm"]["seed_perm"] = seed_perm
    metrics["ClusterPerm"]["t0_cluster"] = float(t0)  # the threshold used
    metrics["ClusterPerm"]["p0_cluster"] = float(p0_cluster)  # if p0_cluster is global

    metrics["ClusterPerm"]["time_cluster_perm_sec"] = float(t_elapsed)

    if len(blob_truth_masks) > 0:
        det = per_blob_cluster_detection(clus, blob_truth_masks, frac=0.10, min_vox=20)
        metrics["ClusterPerm"]["Power_blob_cluster"] = float(det.mean())
        metrics["ClusterPerm"]["Power_all_blobs_cluster"] = float(1.0 if det.size > 0 and det.all() else 0.0)
    else:
        metrics["ClusterPerm"]["Power_blob_cluster"] = np.nan
        metrics["ClusterPerm"]["Power_all_blobs_cluster"] = np.nan



    #perblob detection metrics:
    sig_masks = {"Bonf": bonf, "BH": bh, "BY": by, "BKY": bky, "ClusterPerm": clus}
    for mth, msk in sig_masks.items():
        p_blob, p_all, _ = per_blob_detection_fraction(msk, blob_truth_masks, frac=0.10)
        metrics[mth]["Power_blob"] = float(p_blob) if np.isfinite(p_blob) else np.nan
        metrics[mth]["Power_all_blobs"] = float(p_all) if np.isfinite(p_all) else np.nan

    debug_last = {
        "pmap": pmap,
        "tmap": tmap,
        "t0": float(t0),
        "sig_masks": sig_masks,
    }
    return metrics, debug_last


#visualising our last run shows what each method's detection power is:
def plot_last_run(sc_name, S, truth_mask, debug_last, M=None, z_slice=None):
    tmap = debug_last["tmap"]
    masks = debug_last["sig_masks"]

    Nz_ = S.shape[0]
    if z_slice is None:
        z_slice = Nz_ // 2
    z_slice = int(np.clip(z_slice, 0, Nz_ - 1))

    # Extract 2D slices
    S2 = S[z_slice, :, :]
    T2 = tmap[z_slice, :, :]
    truth2 = truth_mask[z_slice, :, :].astype(bool)

    M2 = None if M is None else M[z_slice, :, :]

    clus2 = masks["ClusterPerm"][z_slice, :, :].astype(bool)
    bonf2 = masks["Bonf"][z_slice, :, :].astype(bool)
    bh2   = masks["BH"][z_slice, :, :].astype(bool)
    by2   = masks["BY"][z_slice, :, :].astype(bool)
    bky2  = masks["BKY"][z_slice, :, :].astype(bool)

    # Plot
    fig, axes = plt.subplots(2, 4, figsize=(16, 8))
    ax = axes.ravel()

    # True signal
    im0 = ax[0].imshow(S2, cmap="coolwarm")
    ax[0].set_title(f"{sc_name}: True signal field S (z={z_slice})")
    ax[0].set_xticks([]); ax[0].set_yticks([])
    if M2 is not None:
        ax[0].contour(M2, levels=[0.5], colors="black", linewidths=1.5)
    ax[0].contour(truth2.astype(float), levels=[0.5], linewidths=1)
    plt.colorbar(im0, ax=ax[0], fraction=0.046, pad=0.04)

    # t-map
    im1 = ax[1].imshow(T2, cmap="coolwarm", vmin=-6, vmax=6)
    ax[1].set_title(f"t-statistic map (one-sided) (z={z_slice})")
    ax[1].set_xticks([]); ax[1].set_yticks([])
    plt.colorbar(im1, ax=ax[1], fraction=0.046, pad=0.04)

    # Truth mask
    ax[2].imshow(truth2, cmap="Greens", alpha=0.85)
    ax[2].set_title("Truth region")
    ax[2].set_xticks([]); ax[2].set_yticks([])

    # Cluster permutation result
    t0_disp = debug_last.get("t0", np.nan)
    ax[3].imshow(clus2, cmap="Reds", alpha=0.85)
    if np.isfinite(t0_disp):
        ax[3].set_title(f"ClusterPerm (t0≈{t0_disp:.2f})")
    else:
        ax[3].set_title("ClusterPerm")
    ax[3].set_xticks([]); ax[3].set_yticks([])

    def show_mask(ax_, mask2d, title):
        ax_.imshow(mask2d, cmap="Reds", alpha=0.85)
        ax_.set_title(title)
        ax_.set_xticks([]); ax_.set_yticks([])

    show_mask(ax[4], bonf2, "Bonf")
    show_mask(ax[5], bh2,   "BH")
    show_mask(ax[6], by2,   "BY")
    show_mask(ax[7], bky2,  "BKY")

    plt.tight_layout()
    plt.show()


#now we can run our scenario
def run_scenario(sc, rng_master):
    sc_name = sc["name"]
    n_subj, fwhm, sigma_N, n_blobs, amp = sc["n_subj"], sc["fwhm"], sc["sigma_N"], sc["n_blobs"], sc["amp"]
    tau = fwhm_to_tau(fwhm)
    centers = sc.get("centers")
    if centers is None:
        centers = get_blob_centers(n_blobs)
    M = None
    df = n_subj - 1
    t0_cluster = float(stats.t.isf(p0_cluster, df=df))
    print(f"[{sc_name}] cluster-form thresh: p0={p0_cluster} (one-sided), df={df}, t0={t0_cluster:.4f}")

    if n_blobs > 0 and amp != 0.0 and len(centers) > 0:
      #creating the binary mask
        M = np.zeros((Nz, Ny, Nx), dtype=float)
        for (x0, y0, z0) in centers:
            M[sphere_mask(x0, y0, z0, blob_radius) & analysis_mask] = 1.0
        S0 = ndimage.gaussian_filter(M,sigma=sigma_S,mode="reflect", cval=0.0)
        S = float(amp) * S0
        S[np.abs(S)<1e-2]=0
        truth_mask = (S > 0.0) & analysis_mask
        true_null_mask = (~truth_mask) & analysis_mask
        truth_n = int(truth_mask.sum())

        blob_truth_masks = []
        for (x0, y0,z0) in centers:
            Mj = (sphere_mask(x0, y0,z0, blob_radius) & analysis_mask).astype(float)
            Sj0 = ndimage.gaussian_filter(Mj, sigma=sigma_S, mode="reflect", cval=0.0)
            Sj0[np.abs(Sj0)<1e-3]=0
            blob_truth_masks.append((Sj0 > 0.0) & analysis_mask)

    else:
        S = np.zeros((Nz, Ny, Nx), dtype=float)
        truth_mask = np.zeros((Nz, Ny, Nx), dtype=bool)
        true_null_mask = analysis_mask.copy()
        blob_truth_masks = []
        truth_n = 0

    store = {m: [] for m in methods}
    runs_rows_local = []
    t0_list = []
    last_debug = None

    for run_idx in range(n_runs):
        seed_run = int(rng_master.integers(0, 2**32 - 1))
        rng_run = np.random.default_rng(seed_run)

        images = simulate_images(S, n_subj=n_subj, sigma_N=sigma_N, tau=tau, rng=rng_run)
        metrics, debug_last = run_one(
            images,
            truth_mask=truth_mask,
            true_null_mask=true_null_mask,
            rng_run=rng_run,
            truth_n=truth_n,
            blob_truth_masks=blob_truth_masks,
            t0_cluster=t0_cluster,
            alpha=alpha,
            q_fdr=q_fdr,
            n_perm_cluster=n_perm_cluster
        )

        t0_list.append(debug_last["t0"])
        last_debug = debug_last

        for mth in methods:
            store[mth].append(metrics[mth])

            row_run = {
                "scenario": sc_name,
                "run": run_idx,
                "seed_run": seed_run,
                "method": mth,
                "n_subj": n_subj,
                "fwhm": fwhm,
                "tau": float(tau),
                "sigma_N": float(sigma_N),
                "n_blobs": n_blobs,
                "amp": float(amp),
                "truth_n": int(truth_n),
                "blob_detect_frac": float(0.10),
                "truth_radius": float(blob_radius),
                "center_sep": sc.get("center_sep", np.nan),
            }
            row_run.update(metrics[mth])
            runs_rows_local.append(row_run)

    return {
        "sc_name": sc_name,
        "store": store,
        "runs_rows": runs_rows_local,
        "t0_list": t0_list,
        "last_debug": last_debug,
        "S": S,
        "truth_mask": truth_mask,
        "true_null_mask": true_null_mask,
        "truth_n": truth_n,
        "tau": tau,
        "M": M,
    }


#we consider different scenarios by changing one hyperparam at a time
scenarios = [
    {"name": "NULL",     "n_subj": 20, "fwhm": 8,  "n_blobs": 0, "amp": 0.0},
    {"name": "Typical",  "n_subj": 20, "fwhm": 8,  "n_blobs": 3, "amp": 2.0},
    {"name": "Lowamp",   "n_subj": 20, "fwhm": 8,  "n_blobs": 3, "amp": 1.5},
    {"name": "Noisy",    "n_subj": 20, "fwhm": 8,  "n_blobs": 3, "amp": 2.0, "target_sd": 2.0},
    {"name": "High_n",   "n_subj": 40, "fwhm": 8,  "n_blobs": 3, "amp": 2.0},
    {"name": "Smooth",   "n_subj": 20, "fwhm": 20, "n_blobs": 3, "amp": 2.0},
    {"name": "NoSmooth", "n_subj": 20, "fwhm": 0,  "n_blobs": 3, "amp": 2.0},
    {"name": "MoreBlobs","n_subj": 20, "fwhm": 8,  "n_blobs": 8, "amp": 2.0},
]
for sc in scenarios:
    target_sd = sc.get("target_sd", post_smooth_sd_target)
    sc["sigma_N"] = sigmaN_for_fwhm(sc["fwhm"], target_sd)

for d in [10, 20, 30, 40, 50]:
    scenarios.append({
        "name": f"Merged_sep{d}",
        "n_subj": 20,
        "center_sep": d,
        "fwhm": 16,
        "n_blobs": 3,
        "amp": 1.0,
        "centers": centers_merged_three(center_sep=d),
        "sigma_N": sigmaN_for_fwhm(16, post_smooth_sd_target),  # ADD THIS
    })

#to check it is correctly calibrated, under the true null p vals should follow an approx unif dist
def plot_pmap_calibration(pmap, true_null_mask, title="p-values", bins=40):
    p = np.asarray(pmap)[true_null_mask].ravel()
    p = p[np.isfinite(p)]
    plt.figure(figsize=(6, 4))
    plt.hist(p, bins=bins, density=True, edgecolor="k")
    plt.axhline(1.0, linestyle="--") # uniform density line
    plt.title(title)
    plt.xlabel("p-value")
    plt.ylabel("density")
    plt.tight_layout()
    plt.show()
#We run the simulation
runs_rows = []
scenario_tables = {}
#to show which one we're on
print(f"Starting sim, n_runs={n_runs}, n_perm_cluster={n_perm_cluster}")
for sc in scenarios:
    res = run_scenario(sc, rng_master)
    grid_n = Nx * Ny * Nz
    truth_n = res["truth_n"]
    print(f"[{res['sc_name']}] truth_n = {truth_n} ({truth_n/grid_n:.1%} of grid), null_n = {grid_n-truth_n}")

    sc_name = res["sc_name"] #go through each scenario
    store = res["store"]
    runs_rows.extend(res["runs_rows"])

    #plots
    plot_last_run(sc_name, res["S"], res["truth_mask"], res["last_debug"], M=res["M"])

    if sc_name == "NULL":
        plot_pmap_calibration(res["last_debug"]["pmap"],res["true_null_mask"], title="NULL p-values (should be ~Uniform)", bins=40)
    #Per scenario metrics
    rows_for_table = []
    is_null_scenario = (sc_name == "NULL")


    for mth in methods:
        df_runs = pd.DataFrame(store[mth])
        n_mc = int(df_runs.shape[0])

        #to check validity
        if "R_gt0" in df_runs.columns:
            k_r = int(np.nansum(df_runs["R_gt0"]))
            pr_R_gt0, pr_R_lo, pr_R_hi = binomial_exact_ci(k_r, n_mc, alpha=0.05)
        else:
            pr_R_gt0 = pr_R_lo = pr_R_hi = np.nan

        mean_R = float(np.nanmean(df_runs["R"])) if "R" in df_runs.columns else np.nan
        med_R = float(np.nanmedian(df_runs["R"]))

        #Voxelwise FWER
        k_fwer = int(np.nansum(df_runs["FWER_voxel_true_null"]))
        fwer_hat, fwer_lo, fwer_hi = binomial_exact_ci(k_fwer, n_mc, alpha=0.05)

        # Clusterwise FWER
        if mth == "ClusterPerm":
            k_cfw = int(np.nansum(df_runs["ClusterFP_any"]))
            cfw_hat, cfw_lo, cfw_hi = binomial_exact_ci(k_cfw, n_mc, alpha=0.05)

            k_any = int(np.nansum(df_runs["ClusterAnySig"]))
            any_hat, any_lo, any_hi = binomial_exact_ci(k_any, n_mc, alpha=0.05)
        else:
            cfw_hat = cfw_lo = cfw_hi = np.nan
            any_hat = any_lo = any_hi = np.nan


        # FDR, we use E[V/max(R,1)]
        if mth in ["BH", "BY", "BKY"]:
            V = df_runs["V_true"].to_numpy(dtype=float)
            R = df_runs["R"].to_numpy(dtype=float)

            fdp_std = df_runs["FDP"].to_numpy(dtype=float)
            fdr_mean, fdr_lo, fdr_hi = bootstrap_ci_mean(fdp_std, seed=3)
        else:
            V = R = None
            fdr_mean = fdr_lo = fdr_hi = np.nan

        #Support performance
        p_mean, p_lo, p_hi = bootstrap_ci_mean(df_runs["Power"], seed=17)
        d_mean, d_lo, d_hi = bootstrap_ci_mean(df_runs["Dice"], seed=2)
        prec = df_runs["Precision"].to_numpy(float)
        prec_mean, prec_lo, prec_hi = bootstrap_ci_mean(prec[np.isfinite(prec)], seed=11)

        ext_mean = float(np.nanmean(df_runs["extent_inflation"])) if np.isfinite(df_runs["extent_inflation"]).any() else np.nan
        if is_null_scenario:
            p_mean = p_lo = p_hi = np.nan
            d_mean = d_lo = d_hi = np.nan
            prec_mean = np.nan
            ext_mean = np.nan
        if is_null_scenario:
            power_blob_mean = np.nan
            power_all_blobs_mean = np.nan
        else:
            power_blob_mean = float(np.nanmean(df_runs["Power_blob"]))
            power_all_blobs_mean = float(np.nanmean(df_runs["Power_all_blobs"]))
        if mth == "ClusterPerm":
            cl_pow_blob = float(np.nanmean(df_runs["Power_blob_cluster"]))
            cl_pow_all = float(np.nanmean(df_runs["Power_all_blobs_cluster"]))
        else:
            cl_pow_blob = np.nan
            cl_pow_all = np.nan

        row = {
            "method": mth,

            #Main metrics
            "Pr(R>0)": pr_R_gt0,
            "mean_R": mean_R,
            "median_R": med_R,

            "FWER_voxel": fwer_hat,
            "FWER_CI_lo": fwer_lo,
            "FWER_CI_hi": fwer_hi,

            "FDR_mean": fdr_mean,
            "FDR_CI_lo": fdr_lo,
            "FDR_CI_hi": fdr_hi,

            "ClusterFP_any": cfw_hat,
            "ClusterFP_CI_lo": cfw_lo,
            "ClusterFP_CI_hi": cfw_hi,

            "ClusterAnySig": any_hat,
            "ClusterAnySig_CI_lo": any_lo,
            "ClusterAnySig_CI_hi": any_hi,



            #to see how truth performs
            "Power": p_mean,
            "Power_CI_lo": p_lo,
            "Power_CI_hi": p_hi,

            "Dice": d_mean,
            "Dice_CI_lo": d_lo,
            "Dice_CI_hi": d_hi,

            "Precision": prec_mean,
            "Precision_CI_lo": prec_lo,
            "Precision_CI_hi": prec_hi,

            "extent_inflation": ext_mean,

            "Power_blob": power_blob_mean,
            "Power_all_blobs": power_all_blobs_mean,
            "Power_blob_cluster": cl_pow_blob,
            "Power_all_blobs_cluster": cl_pow_all,


            "Pr(R>0)_CI_lo": pr_R_lo,
            "Pr(R>0)_CI_hi": pr_R_hi,

            #summary of flse pos
            "V_true_mean": float(np.nanmean(df_runs["V_true"])) if "V_true" in df_runs.columns else np.nan,
        }

        rows_for_table.append(row)

    scenario_table = pd.DataFrame(rows_for_table).round(3).set_index("method")
    scenario_tables[sc_name] = scenario_table

    print(f"\n--- summary table: {sc_name} ---")
    print(scenario_table.to_string())

    scenario_table.to_csv(os.path.join(OUT_DIR, f"scenario_summary_{sc_name}.csv"), index=True)

runs_df = pd.DataFrame(runs_rows)
key_cols = ["FDP", "FDP_cond_Rgt0",
            "FWER_voxel_true_null", "R", "Power", "Dice",
            "ClusterFP_any", "ClusterAnySig"]


for col in key_cols:
    if col in runs_df.columns:
        tmp = runs_df[["scenario", "method", "run", col]].copy()
        tmp = add_running_stats(tmp, value_col=col, group_cols=["scenario", "method"])
        runs_df = runs_df.merge(
            tmp[["scenario", "method", "run", f"{col}_running_mean", f"{col}_running_se"]],
            on=["scenario", "method", "run"],
            how="left"
        )

runs_path = os.path.join(OUT_DIR, "simulation_per_run_metrics.csv")
runs_df.to_csv(runs_path, index=False)
print("\nSaved:", runs_path)

runs_df.to_pickle(os.path.join(OUT_DIR, "simulation_per_run_metrics.pkl"))
print("Saved pickle too.")

print("\nSaved simulation_per_run_metrics.csv")
```

```python
import os
import numpy as np
import nibabel as nib
from scipy import ndimage, stats

from nilearn import datasets, image
from nilearn.glm.first_level import FirstLevelModel


# =========================
# Settings
# =========================
OUT_DIR = "haxby_3d_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

alpha = 0.05
q_fdr = 0.05

p0_cluster = 0.01
n_perm_cluster = 1500
chunk = 64

STRUCT_3D = np.ones((3, 3, 3), dtype=int)
_SIGNS = np.array([-1.0, 1.0], dtype=float)


# =========================
# labels -> events
# =========================
def labels_to_events(labels, tr):
    import pandas as pd
    labels = np.asarray(labels, dtype=str)
    labels = np.char.lower(np.char.strip(labels))

    onsets, durations, types = [], [], []
    n = labels.size
    start = 0
    while start < n:
        lab = labels[start]
        end = start + 1
        while end < n and labels[end] == lab:
            end += 1
        onsets.append(start * tr)
        durations.append((end - start) * tr)
        types.append(lab)
        start = end

    return pd.DataFrame({"onset": onsets, "duration": durations, "trial_type": types})


# =========================
# Multiple comparisons
# =========================
def bonferroni(pmap, mask_bool, alpha):
    p = pmap[mask_bool]
    p = p[np.isfinite(p)]
    m = int(p.size)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if m > 0:
        sig[mask_bool] = pmap[mask_bool] <= (alpha / m)
    return sig

def stepup_fdr(pmap, mask_bool, q, mode="BH"):
    p = pmap[mask_bool]
    p = p[np.isfinite(p)]
    m = int(p.size)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if m == 0:
        return sig, np.nan

    p_sorted = np.sort(p)
    i = np.arange(1, m + 1, dtype=float)

    if mode == "BH":
        line = (i / m) * q
        ok = p_sorted <= line
        pcrit = p_sorted[np.max(np.where(ok)[0])] if np.any(ok) else np.nan

    elif mode == "BY":
        c_m = np.sum(1.0 / np.arange(1, m + 1, dtype=float))
        line = (i / m) * (q / c_m)
        ok = p_sorted <= line
        pcrit = p_sorted[np.max(np.where(ok)[0])] if np.any(ok) else np.nan

    elif mode == "BKY":
        q0 = q / (1.0 + q)
        line1 = (i / m) * q0
        ok1 = p_sorted <= line1
        pcrit1 = p_sorted[np.max(np.where(ok1)[0])] if np.any(ok1) else np.nan

        R1 = int(np.sum(p <= pcrit1)) if np.isfinite(pcrit1) else 0
        m0_hat = max(1, m - R1)
        q_star = min(1.0, q0 * (m / m0_hat))

        line2 = (i / m) * q_star
        ok2 = p_sorted <= line2
        pcrit = p_sorted[np.max(np.where(ok2)[0])] if np.any(ok2) else np.nan
    else:
        raise ValueError("mode must be 'BH', 'BY', or 'BKY'")

    if np.isfinite(pcrit):
        sig[mask_bool] = pmap[mask_bool] <= pcrit
    return sig, float(pcrit) if np.isfinite(pcrit) else np.nan


# =========================
# Group voxelwise one-sample t + one-sided p
# =========================
def compute_group_t_p(images_4d, mask_bool, eps=1e-8):
    """
    images_4d: (n, X, Y, Z)
    returns tmap, pmap (one-sided)
    """
    images_4d = np.asarray(images_4d, dtype=float)
    n = images_4d.shape[0]
    df = n - 1

    tmap = np.full(mask_bool.shape, np.nan, dtype=float)
    pmap = np.full(mask_bool.shape, np.nan, dtype=float)

    Xv = images_4d[:, mask_bool]
    mean = Xv.mean(axis=0)
    sd = Xv.std(axis=0, ddof=1)
    tvals = mean / np.maximum(sd / np.sqrt(n), eps)
    pvals = stats.t.sf(tvals, df=df)

    tmap[mask_bool] = tvals
    pmap[mask_bool] = np.clip(pvals, 1e-300, 1.0)
    return tmap, pmap


# =========================
# 3D cluster permutation
# =========================
def cluster_perm_onesample_one_sided_3d(images_4d, t0, alpha, n_perm, mask_bool, rng,
                                        structure=STRUCT_3D, eps=1e-12, chunk=64):
    images_4d = np.asarray(images_4d, dtype=float)
    n = images_4d.shape[0]
    df = n - 1

    V = int(mask_bool.sum())
    Xv = images_4d[:, mask_bool]  # (n, V)
    SS = np.sum(Xv**2, axis=0)

    # observed
    mean_obs = Xv.mean(axis=0)
    var_obs = (SS - n * mean_obs**2) / max(df, 1)
    var_obs = np.maximum(var_obs, 0.0)
    denom_obs = np.sqrt(var_obs / n) + eps
    t_obs_v = mean_obs / denom_obs

    t_obs = np.full(mask_bool.shape, -np.inf, dtype=float)
    t_obs[mask_bool] = t_obs_v

    # null max cluster sizes
    null_max = np.zeros(n_perm, dtype=int)
    signs_all = rng.choice(_SIGNS, size=(n_perm, n), replace=True)

    for start in range(0, n_perm, chunk):
        stop = min(n_perm, start + chunk)
        signs = signs_all[start:stop]
        mean_perm = (signs @ Xv) / n
        var_perm = (SS[None, :] - n * mean_perm**2) / max(df, 1)
        var_perm = np.maximum(var_perm, 0.0)
        denom_perm = np.sqrt(var_perm / n) + eps
        t_perm = mean_perm / denom_perm

        for b in range(stop - start):
            tmap_b = np.full(mask_bool.shape, -np.inf, dtype=float)
            tmap_b[mask_bool] = t_perm[b]
            supra = (tmap_b >= t0) & mask_bool
            lab, nlab = ndimage.label(supra, structure=structure)
            if nlab == 0:
                null_max[start + b] = 0
            else:
                counts = np.bincount(lab.ravel())
                null_max[start + b] = int(counts[1:].max(initial=0))

    supra_obs = (t_obs >= t0) & mask_bool
    lab, nlab = ndimage.label(supra_obs, structure=structure)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if nlab == 0:
        return sig

    sizes_obs = np.bincount(lab.ravel())[1:]
    ge = (null_max[:, None] >= sizes_obs[None, :])
    pvals = (1.0 + ge.sum(axis=0)) / (n_perm + 1.0)

    for k, p in enumerate(pvals, start=1):
        if p <= alpha:
            sig[lab == k] = True

    return sig


# =========================
# Save helper
# =========================
def save_img(data3d, ref_img, name, dtype=np.float32):
    path = os.path.join(OUT_DIR, f"{name}.nii.gz")
    nib.save(nib.Nifti1Image(data3d.astype(dtype), ref_img.affine, header=ref_img.header), path)
    print("Saved:", path)
    return path


# =========================
# Fetch Haxby
# =========================
print("Fetching Haxby...")
haxby = datasets.fetch_haxby()  # colab-safe API
print("Keys:", list(haxby.keys()))
print("n subjects:", len(haxby.func))

TR = float(getattr(haxby, "t_r", 2.5))  # many builds include t_r; else fallback
print("TR:", TR)

# We'll use VT mask to keep it fast
base_mask_img = nib.load(haxby.mask_vt[0])

# =========================
# First-level contrasts
# =========================
contrasts = []
ref_img_for_saving = None

for subj_idx, (func_path, labels_path) in enumerate(zip(haxby.func, haxby.session_target), start=1):
    print(f"\nSubject {subj_idx}")
    print("  func  :", func_path)
    print("  labels:", labels_path)

    # Load functional now so we can resample mask into its grid (avoids sform warnings)
    func_img = nib.load(func_path)
    if ref_img_for_saving is None:
        ref_img_for_saving = func_img.slicer[..., 0]  # 3D reference for saving

    # Resample mask to this functional space once
    mask_img = image.resample_to_img(base_mask_img, ref_img_for_saving, interpolation="nearest")
    mask_bool = mask_img.get_fdata().astype(bool)

    # Load only first column (condition name)
    labels = np.loadtxt(labels_path, dtype=str, usecols=0)
    events = labels_to_events(labels, tr=TR)

    fm = FirstLevelModel(
        t_r=TR,
        mask_img=mask_img,
        hrf_model="spm",
        drift_model="cosine",
        high_pass=1/128,
        noise_model="ar1",
        standardize=False,
        minimize_memory=True,
    )
    fm = fm.fit(func_img, events=events)

    dm = fm.design_matrices_[0]
    cols_lower = [c.lower() for c in dm.columns.tolist()]

    def find_col(prefixes):
        for pref in prefixes:
            for j, c in enumerate(cols_lower):
                if c == pref or c.startswith(pref):
                    return j, dm.columns.tolist()[j]
        return None, None

    j_face, col_face = find_col(["face", "faces"])
    j_house, col_house = find_col(["house", "houses"])

    if j_face is None or j_house is None:
        print("  WARNING: could not find face/house regressors. Skipping.")
        print("  Columns (first 25):", dm.columns.tolist()[:25])
        continue

    print(f"  Using columns: face='{col_face}', house='{col_house}'")

    con = np.zeros(len(dm.columns), dtype=float)
    con[j_face] = 1.0
    con[j_house] = -1.0

    con_img = fm.compute_contrast(con, output_type="effect_size")
    # con_img is already in the same mask/func space (no need to resample again)
    contrasts.append(con_img)

print(f"\nBuilt {len(contrasts)} usable subject contrast maps.")


# =========================
# If >=3 subjects: group analysis + permutation
# If <3 subjects: single-subject output + voxelwise corrections only
# =========================
if len(contrasts) >= 3:
    # Build stack
    # Use the mask space used for the first subject
    mask_img = image.resample_to_img(base_mask_img, ref_img_for_saving, interpolation="nearest")
    mask_bool = mask_img.get_fdata().astype(bool)

    images_4d = np.stack([img.get_fdata().astype(float) for img in contrasts], axis=0)
    tmap, pmap = compute_group_t_p(images_4d, mask_bool)

    sig_bonf = bonferroni(pmap, mask_bool, alpha)
    sig_bh,  pcrit_bh  = stepup_fdr(pmap, mask_bool, q_fdr, mode="BH")
    sig_by,  pcrit_by  = stepup_fdr(pmap, mask_bool, q_fdr, mode="BY")
    sig_bky, pcrit_bky = stepup_fdr(pmap, mask_bool, q_fdr, mode="BKY")

    df = images_4d.shape[0] - 1
    t0_cluster = float(stats.t.isf(p0_cluster, df=df))
    rng_perm = np.random.default_rng(0)
    sig_cluster = cluster_perm_onesample_one_sided_3d(
        images_4d, t0=t0_cluster, alpha=alpha,
        n_perm=n_perm_cluster, mask_bool=mask_bool,
        rng=rng_perm, structure=STRUCT_3D, chunk=chunk
    )

    print("\nThresholds:")
    print("  BH  pcrit:", pcrit_bh)
    print("  BY  pcrit:", pcrit_by)
    print("  BKY pcrit:", pcrit_bky)
    print("  cluster t0:", t0_cluster)

    save_img(tmap, ref_img_for_saving, "group_tmap_faces_minus_houses", np.float32)
    save_img(pmap, ref_img_for_saving, "group_pmap_faces_minus_houses", np.float32)
    save_img(sig_bonf.astype(np.uint8), ref_img_for_saving, "sig_Bonf", np.uint8)
    save_img(sig_bh.astype(np.uint8),   ref_img_for_saving, "sig_BH", np.uint8)
    save_img(sig_by.astype(np.uint8),   ref_img_for_saving, "sig_BY", np.uint8)
    save_img(sig_bky.astype(np.uint8),  ref_img_for_saving, "sig_BKY", np.uint8)
    save_img(sig_cluster.astype(np.uint8), ref_img_for_saving, "sig_ClusterPerm", np.uint8)

    print("\nDone. Outputs in:", OUT_DIR)

else:
    # Single-subject fallback
    print("\nOnly", len(contrasts), "subject available in this nilearn/Haxby build.")
    print("Running single-subject map + voxelwise corrections only (Bonf/BH/BY/BKY).")
    print("Group t-test + cluster permutation require >= 3 subjects.")

    # We'll treat the single subject contrast values as 'effect' and produce an uncorrected z-ish map is not valid.
    # Instead, we just threshold its contrast *pmap not defined*. So we output the contrast itself and skip t/p.
    # If you want single-subject inference, you need the subject-level t-stat map from the first-level model.

    # Save the single subject effect-size contrast
    con_img = contrasts[0]
    con_data = con_img.get_fdata().astype(np.float32)
    save_img(con_data, ref_img_for_saving, "subject1_effect_size_faces_minus_houses", np.float32)

    print("\nSaved single-subject contrast effect-size map.")
    print("To do inference at single-subject level, switch output_type='stat' in compute_contrast and use that t-map.")
```

```python
import os
import numpy as np
import nibabel as nib
from scipy import ndimage, stats

from nilearn import datasets, image
from nilearn.glm.first_level import FirstLevelModel

OUT_DIR = "haxby_3d_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

alpha = 0.05
q_fdr = 0.05

# cluster params (only used if n_subj >= 3)
p0_cluster = 0.01
n_perm_cluster = 1500
chunk = 64
STRUCT_3D = np.ones((3, 3, 3), dtype=int)
_SIGNS = np.array([-1.0, 1.0], dtype=float)


# -------------------------
# helpers
# -------------------------
def labels_to_events(labels, tr):
    import pandas as pd
    labels = np.asarray(labels, dtype=str)
    labels = np.char.lower(np.char.strip(labels))

    onsets, durations, types = [], [], []
    n = labels.size
    start = 0
    while start < n:
        lab = labels[start]
        end = start + 1
        while end < n and labels[end] == lab:
            end += 1
        onsets.append(start * tr)
        durations.append((end - start) * tr)
        types.append(lab)
        start = end
    return pd.DataFrame({"onset": onsets, "duration": durations, "trial_type": types})


def bonferroni(pmap, mask_bool, alpha):
    p = pmap[mask_bool]
    p = p[np.isfinite(p)]
    m = int(p.size)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if m > 0:
        sig[mask_bool] = pmap[mask_bool] <= (alpha / m)
    return sig


def stepup_fdr(pmap, mask_bool, q, mode="BH"):
    p = pmap[mask_bool]
    p = p[np.isfinite(p)]
    m = int(p.size)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if m == 0:
        return sig, np.nan

    p_sorted = np.sort(p)
    i = np.arange(1, m + 1, dtype=float)

    if mode == "BH":
        line = (i / m) * q
        ok = p_sorted <= line
        pcrit = p_sorted[np.max(np.where(ok)[0])] if np.any(ok) else np.nan

    elif mode == "BY":
        c_m = np.sum(1.0 / np.arange(1, m + 1, dtype=float))
        line = (i / m) * (q / c_m)
        ok = p_sorted <= line
        pcrit = p_sorted[np.max(np.where(ok)[0])] if np.any(ok) else np.nan

    elif mode == "BKY":
        q0 = q / (1.0 + q)
        line1 = (i / m) * q0
        ok1 = p_sorted <= line1
        pcrit1 = p_sorted[np.max(np.where(ok1)[0])] if np.any(ok1) else np.nan

        R1 = int(np.sum(p <= pcrit1)) if np.isfinite(pcrit1) else 0
        m0_hat = max(1, m - R1)
        q_star = min(1.0, q0 * (m / m0_hat))

        line2 = (i / m) * q_star
        ok2 = p_sorted <= line2
        pcrit = p_sorted[np.max(np.where(ok2)[0])] if np.any(ok2) else np.nan
    else:
        raise ValueError("mode must be 'BH', 'BY', or 'BKY'")

    if np.isfinite(pcrit):
        sig[mask_bool] = pmap[mask_bool] <= pcrit
    return sig, float(pcrit) if np.isfinite(pcrit) else np.nan


def compute_group_t_p(images_4d, mask_bool, eps=1e-8):
    images_4d = np.asarray(images_4d, dtype=float)
    n = images_4d.shape[0]
    df = n - 1

    tmap = np.full(mask_bool.shape, np.nan, dtype=float)
    pmap = np.full(mask_bool.shape, np.nan, dtype=float)

    Xv = images_4d[:, mask_bool]
    mean = Xv.mean(axis=0)
    sd = Xv.std(axis=0, ddof=1)
    tvals = mean / np.maximum(sd / np.sqrt(n), eps)
    pvals = stats.t.sf(tvals, df=df)  # one-sided

    tmap[mask_bool] = tvals
    pmap[mask_bool] = np.clip(pvals, 1e-300, 1.0)
    return tmap, pmap


def cluster_perm_onesample_one_sided_3d(images_4d, t0, alpha, n_perm, mask_bool, rng,
                                        structure=STRUCT_3D, eps=1e-12, chunk=64):
    images_4d = np.asarray(images_4d, dtype=float)
    n = images_4d.shape[0]
    df = n - 1

    V = int(mask_bool.sum())
    Xv = images_4d[:, mask_bool]
    SS = np.sum(Xv**2, axis=0)

    mean_obs = Xv.mean(axis=0)
    var_obs = (SS - n * mean_obs**2) / max(df, 1)
    var_obs = np.maximum(var_obs, 0.0)
    denom_obs = np.sqrt(var_obs / n) + eps
    t_obs_v = mean_obs / denom_obs

    t_obs = np.full(mask_bool.shape, -np.inf, dtype=float)
    t_obs[mask_bool] = t_obs_v

    null_max = np.zeros(n_perm, dtype=int)
    signs_all = rng.choice(_SIGNS, size=(n_perm, n), replace=True)

    for start in range(0, n_perm, chunk):
        stop = min(n_perm, start + chunk)
        signs = signs_all[start:stop]

        mean_perm = (signs @ Xv) / n
        var_perm = (SS[None, :] - n * mean_perm**2) / max(df, 1)
        var_perm = np.maximum(var_perm, 0.0)
        denom_perm = np.sqrt(var_perm / n) + eps
        t_perm = mean_perm / denom_perm

        for b in range(stop - start):
            tmap_b = np.full(mask_bool.shape, -np.inf, dtype=float)
            tmap_b[mask_bool] = t_perm[b]
            supra = (tmap_b >= t0) & mask_bool
            lab, nlab = ndimage.label(supra, structure=structure)
            if nlab == 0:
                null_max[start + b] = 0
            else:
                counts = np.bincount(lab.ravel())
                null_max[start + b] = int(counts[1:].max(initial=0))

    supra_obs = (t_obs >= t0) & mask_bool
    lab, nlab = ndimage.label(supra_obs, structure=structure)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if nlab == 0:
        return sig

    sizes_obs = np.bincount(lab.ravel())[1:]
    ge = (null_max[:, None] >= sizes_obs[None, :])
    pvals = (1.0 + ge.sum(axis=0)) / (n_perm + 1.0)

    for k, p in enumerate(pvals, start=1):
        if p <= alpha:
            sig[lab == k] = True
    return sig


def save_like(ref_img, data3d, name, dtype=np.float32):
    path = os.path.join(OUT_DIR, f"{name}.nii.gz")
    out = nib.Nifti1Image(data3d.astype(dtype), ref_img.affine, header=ref_img.header)
    nib.save(out, path)
    print("Saved:", path)
    return path


# -------------------------
# Fetch Haxby
# -------------------------
print("Fetching Haxby...")
haxby = datasets.fetch_haxby()
print("Keys:", list(haxby.keys()))
print("n subjects:", len(haxby.func))

TR = float(getattr(haxby, "t_r", 2.5))
print("TR:", TR)

# We'll use VT mask (fast)
base_mask_img = nib.load(haxby.mask_vt[0])

# -------------------------
# First-level per subject
# -------------------------
subject_effect_imgs = []
subject_stat_imgs = []

for subj_idx, (func_path, labels_path) in enumerate(zip(haxby.func, haxby.session_target), start=1):
    print(f"\nSubject {subj_idx}")
    print("  func  :", func_path)
    print("  labels:", labels_path)

    func_img = nib.load(func_path)
    ref_3d = func_img.slicer[..., 0]  # reference grid for this subject

    # resample mask ONCE to functional grid (nearest keeps it binary)
    mask_img = image.resample_to_img(base_mask_img, ref_3d, interpolation="nearest")
    mask_bool = mask_img.get_fdata().astype(bool)

    labels = np.loadtxt(labels_path, dtype=str, usecols=0)
    events = labels_to_events(labels, tr=TR)

    fm = FirstLevelModel(
        t_r=TR,
        mask_img=mask_img,
        hrf_model="spm",
        drift_model="cosine",
        high_pass=1/128,
        noise_model="ar1",
        standardize=False,
        minimize_memory=True,
    )
    fm = fm.fit(func_img, events=events)

    dm = fm.design_matrices_[0]
    cols_lower = [c.lower() for c in dm.columns.tolist()]

    def find_col(prefixes):
        for pref in prefixes:
            for j, c in enumerate(cols_lower):
                if c == pref or c.startswith(pref):
                    return j, dm.columns.tolist()[j]
        return None, None

    j_face, col_face = find_col(["face", "faces"])
    j_house, col_house = find_col(["house", "houses"])

    if j_face is None or j_house is None:
        print("  WARNING: could not find face/house regressors. Skipping.")
        print("  Columns (first 25):", dm.columns.tolist()[:25])
        continue

    print(f"  Using columns: face='{col_face}', house='{col_house}'")

    con = np.zeros(len(dm.columns), dtype=float)
    con[j_face] = 1.0
    con[j_house] = -1.0

    # Save BOTH: effect size (beta-ish) and stat (t-map)
    eff_img = fm.compute_contrast(con, output_type="effect_size")
    stat_img = fm.compute_contrast(con, output_type="stat")  # <-- t-map

    subject_effect_imgs.append(eff_img)
    subject_stat_imgs.append(stat_img)

print(f"\nBuilt {len(subject_stat_imgs)} usable subject maps.")


# -------------------------
# Case 1: Single subject (n=1) -> valid single-subject inference using t-map
# -------------------------
if len(subject_stat_imgs) == 1:
    print("\nSingle-subject mode: running voxelwise corrections on subject t-map p-values.")
    stat_img = subject_stat_imgs[0]
    t_data = stat_img.get_fdata().astype(float)

    # Mask should match this subject's grid (recompute)
    ref_3d = nib.load(haxby.func[0]).slicer[..., 0]
    mask_img = image.resample_to_img(base_mask_img, ref_3d, interpolation="nearest")
    mask_bool = mask_img.get_fdata().astype(bool)

    # First-level t-map df is internal; nilearn computes it, but doesn't expose df easily.
    # We will use a NORMAL approximation for voxelwise p-values here (common in first-level maps),
    # OR you can output the t-map and let external software do df-specific p-values.
    # If you want df-specific voxelwise p, tell me and I’ll show how to extract df from the model.
    p_data = np.full(t_data.shape, np.nan, dtype=float)
    p_data[mask_bool] = stats.norm.sf(t_data[mask_bool])  # one-sided approx

    sig_bonf = bonferroni(p_data, mask_bool, alpha)
    sig_bh,  pcrit_bh  = stepup_fdr(p_data, mask_bool, q_fdr, mode="BH")
    sig_by,  pcrit_by  = stepup_fdr(p_data, mask_bool, q_fdr, mode="BY")
    sig_bky, pcrit_bky = stepup_fdr(p_data, mask_bool, q_fdr, mode="BKY")

    print("\nSingle-subject thresholds:")
    print("  BH  pcrit:", pcrit_bh)
    print("  BY  pcrit:", pcrit_by)
    print("  BKY pcrit:", pcrit_bky)

    save_like(ref_3d, subject_effect_imgs[0].get_fdata(), "sub1_effect_faces_minus_houses", np.float32)
    save_like(ref_3d, t_data, "sub1_tmap_faces_minus_houses", np.float32)
    save_like(ref_3d, p_data, "sub1_pmap_faces_minus_houses_normapprox", np.float32)

    save_like(ref_3d, sig_bonf.astype(np.uint8), "sub1_sig_Bonf", np.uint8)
    save_like(ref_3d, sig_bh.astype(np.uint8),   "sub1_sig_BH", np.uint8)
    save_like(ref_3d, sig_by.astype(np.uint8),   "sub1_sig_BY", np.uint8)
    save_like(ref_3d, sig_bky.astype(np.uint8),  "sub1_sig_BKY", np.uint8)

    print("\nDone. Outputs in:", OUT_DIR)
    print("Note: Cluster permutation is not run because it requires >=3 subjects.")

# -------------------------
# Case 2: Group mode (n>=3) -> full pipeline + permutation
# -------------------------
elif len(subject_effect_imgs) >= 3:
    print("\nGroup mode: running group one-sample t + voxelwise corrections + cluster permutation.")

    # Use the first subject’s reference grid for group stacking
    ref_3d = nib.load(haxby.func[0]).slicer[..., 0]
    mask_img = image.resample_to_img(base_mask_img, ref_3d, interpolation="nearest")
    mask_bool = mask_img.get_fdata().astype(bool)

    # Stack subject effect-size contrasts for group one-sample test
    images_4d = np.stack([img.get_fdata().astype(float) for img in subject_effect_imgs], axis=0)
    tmap, pmap = compute_group_t_p(images_4d, mask_bool)

    sig_bonf = bonferroni(pmap, mask_bool, alpha)
    sig_bh,  pcrit_bh  = stepup_fdr(pmap, mask_bool, q_fdr, mode="BH")
    sig_by,  pcrit_by  = stepup_fdr(pmap, mask_bool, q_fdr, mode="BY")
    sig_bky, pcrit_bky = stepup_fdr(pmap, mask_bool, q_fdr, mode="BKY")

    df = images_4d.shape[0] - 1
    t0_cluster = float(stats.t.isf(p0_cluster, df=df))
    rng_perm = np.random.default_rng(0)
    sig_cluster = cluster_perm_onesample_one_sided_3d(
        images_4d, t0=t0_cluster, alpha=alpha, n_perm=n_perm_cluster,
        mask_bool=mask_bool, rng=rng_perm, structure=STRUCT_3D, chunk=chunk
    )

    print("\nGroup thresholds:")
    print("  BH  pcrit:", pcrit_bh)
    print("  BY  pcrit:", pcrit_by)
    print("  BKY pcrit:", pcrit_bky)
    print("  cluster t0:", t0_cluster)

    save_like(ref_3d, tmap, "group_tmap_faces_minus_houses", np.float32)
    save_like(ref_3d, pmap, "group_pmap_faces_minus_houses", np.float32)

    save_like(ref_3d, sig_bonf.astype(np.uint8), "group_sig_Bonf", np.uint8)
    save_like(ref_3d, sig_bh.astype(np.uint8),   "group_sig_BH", np.uint8)
    save_like(ref_3d, sig_by.astype(np.uint8),   "group_sig_BY", np.uint8)
    save_like(ref_3d, sig_bky.astype(np.uint8),  "group_sig_BKY", np.uint8)
    save_like(ref_3d, sig_cluster.astype(np.uint8), "group_sig_ClusterPerm", np.uint8)

    print("\nDone. Outputs in:", OUT_DIR)

else:
    raise RuntimeError(
        f"Need >=1 usable subject map. Built {len(subject_stat_imgs)}. "
        "If 0, face/house labels likely don’t match."
    )
```

```python
# --- If needed ---
# !pip -q install nilearn nibabel scipy pandas scikit-learn matplotlib

import os
import shutil
import numpy as np
import pandas as pd
import nibabel as nib
from scipy import ndimage, stats

from nilearn import datasets, image, plotting
from nilearn.glm.first_level import FirstLevelModel

# =========================
# Settings
# =========================
OUT_DIR = "haxby_3d_outputs_group"
os.makedirs(OUT_DIR, exist_ok=True)

alpha = 0.05
q_fdr = 0.05

p0_cluster = 0.01        # cluster-forming p (one-sided)
n_perm_cluster = 1000    # raise to 1500/5000 later
chunk = 64

STRUCT_3D = np.ones((3, 3, 3), dtype=int)
_SIGNS = np.array([-1.0, 1.0], dtype=float)

TR = 2.5  # Haxby TR

# =========================
# IMPORTANT: Force fresh multi-subject Haxby download
# =========================
# Put Haxby in a NEW directory so you don't reuse the old single-subject cache
DATA_DIR = "/content/nilearn_haxby_full"
if os.path.exists(DATA_DIR):
    shutil.rmtree(DATA_DIR)

print("Fetching Haxby subjects 1..6 into:", DATA_DIR)
haxby = datasets.fetch_haxby(data_dir=DATA_DIR, subjects=[1, 2, 3, 4, 5, 6])
print("n subjects fetched:", len(haxby.func))
print("Example func:", haxby.func[0])

if len(haxby.func) < 3:
    raise RuntimeError(
        f"Still only got {len(haxby.func)} subjects. "
        "This should be 6 if the download succeeded."
    )

# =========================
# Helpers
# =========================
def labels_to_events(labels_1d, tr):
    import pandas as pd
    labels = np.asarray(labels_1d, dtype=str)
    labels = np.char.lower(np.char.strip(labels))

    onsets, durations, types = [], [], []
    n = labels.size
    start = 0
    while start < n:
        lab = labels[start]
        end = start + 1
        while end < n and labels[end] == lab:
            end += 1
        onsets.append(start * tr)
        durations.append((end - start) * tr)
        types.append(lab)
        start = end

    return pd.DataFrame({"onset": onsets, "duration": durations, "trial_type": types})

def compute_group_t_p(images_4d, mask_bool, eps=1e-8):
    images_4d = np.asarray(images_4d, dtype=float)
    n = images_4d.shape[0]
    df = n - 1

    tmap = np.full(mask_bool.shape, np.nan, dtype=float)
    pmap = np.full(mask_bool.shape, np.nan, dtype=float)

    Xv = images_4d[:, mask_bool]
    mean = Xv.mean(axis=0)
    sd = Xv.std(axis=0, ddof=1)

    tvals = mean / np.maximum(sd / np.sqrt(n), eps)
    pvals = stats.t.sf(tvals, df=df)  # one-sided

    tmap[mask_bool] = tvals
    pmap[mask_bool] = np.clip(pvals, 1e-300, 1.0)
    return tmap, pmap

def bonferroni(pmap, mask_bool, alpha):
    p = pmap[mask_bool]
    p = p[np.isfinite(p)]
    m = int(p.size)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if m > 0:
        sig[mask_bool] = pmap[mask_bool] <= (alpha / m)
    return sig

def stepup_fdr(pmap, mask_bool, q, mode="BH"):
    p = pmap[mask_bool]
    p = p[np.isfinite(p)]
    m = int(p.size)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if m == 0:
        return sig, np.nan

    p_sorted = np.sort(p)
    i = np.arange(1, m + 1, dtype=float)

    if mode == "BH":
        line = (i / m) * q
        ok = p_sorted <= line
        pcrit = p_sorted[np.max(np.where(ok)[0])] if np.any(ok) else np.nan

    elif mode == "BY":
        c_m = np.sum(1.0 / np.arange(1, m + 1, dtype=float))
        line = (i / m) * (q / c_m)
        ok = p_sorted <= line
        pcrit = p_sorted[np.max(np.where(ok)[0])] if np.any(ok) else np.nan

    elif mode == "BKY":
        q0 = q / (1.0 + q)
        line1 = (i / m) * q0
        ok1 = p_sorted <= line1
        pcrit1 = p_sorted[np.max(np.where(ok1)[0])] if np.any(ok1) else np.nan

        R1 = int(np.sum(p <= pcrit1)) if np.isfinite(pcrit1) else 0
        m0_hat = max(1, m - R1)
        q_star = min(1.0, q0 * (m / m0_hat))

        line2 = (i / m) * q_star
        ok2 = p_sorted <= line2
        pcrit = p_sorted[np.max(np.where(ok2)[0])] if np.any(ok2) else np.nan
    else:
        raise ValueError("mode must be 'BH', 'BY', or 'BKY'")

    if np.isfinite(pcrit):
        sig[mask_bool] = pmap[mask_bool] <= pcrit

    return sig, float(pcrit) if np.isfinite(pcrit) else np.nan

def cluster_perm_onesample_one_sided_3d(images_4d, t0, alpha, n_perm, mask_bool, rng,
                                        structure=STRUCT_3D, eps=1e-12, chunk=64):
    images_4d = np.asarray(images_4d, dtype=float)
    n = images_4d.shape[0]
    df = n - 1

    V = int(mask_bool.sum())
    Xv = images_4d[:, mask_bool]
    SS = np.sum(Xv**2, axis=0)

    mean_obs = Xv.mean(axis=0)
    var_obs = (SS - n * mean_obs**2) / max(df, 1)
    var_obs = np.maximum(var_obs, 0.0)
    denom_obs = np.sqrt(var_obs / n) + eps
    t_obs_v = mean_obs / denom_obs

    t_obs = np.full(mask_bool.shape, -np.inf, dtype=float)
    t_obs[mask_bool] = t_obs_v

    null_max = np.zeros(n_perm, dtype=int)
    signs_all = rng.choice(_SIGNS, size=(n_perm, n), replace=True)

    for start in range(0, n_perm, chunk):
        stop = min(n_perm, start + chunk)
        signs = signs_all[start:stop]

        mean_perm = (signs @ Xv) / n
        var_perm = (SS[None, :] - n * mean_perm**2) / max(df, 1)
        var_perm = np.maximum(var_perm, 0.0)
        denom_perm = np.sqrt(var_perm / n) + eps
        t_perm = mean_perm / denom_perm

        for b in range(stop - start):
            tmap_b = np.full(mask_bool.shape, -np.inf, dtype=float)
            tmap_b[mask_bool] = t_perm[b]
            supra = (tmap_b >= t0) & mask_bool
            lab, nlab = ndimage.label(supra, structure=structure)
            if nlab == 0:
                null_max[start + b] = 0
            else:
                counts = np.bincount(lab.ravel())
                null_max[start + b] = int(counts[1:].max(initial=0))

    supra_obs = (t_obs >= t0) & mask_bool
    lab, nlab = ndimage.label(supra_obs, structure=structure)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if nlab == 0:
        return sig

    sizes_obs = np.bincount(lab.ravel())[1:]
    ge = (null_max[:, None] >= sizes_obs[None, :])
    pvals = (1.0 + ge.sum(axis=0)) / (n_perm + 1.0)

    for k, p in enumerate(pvals, start=1):
        if p <= alpha:
            sig[lab == k] = True
    return sig

def cluster_stats(sig_mask, mask_bool, structure=STRUCT_3D):
    sig = sig_mask.astype(bool) & mask_bool
    lab, nlab = ndimage.label(sig, structure=structure)
    if nlab == 0:
        return 0, np.nan, np.nan, 0
    sizes = np.bincount(lab.ravel())[1:]
    return int(nlab), float(np.mean(sizes)), float(np.median(sizes)), int(np.max(sizes))

def save_like(ref_img, data3d, name, dtype=np.float32):
    path = os.path.join(OUT_DIR, f"{name}.nii.gz")
    nib.save(nib.Nifti1Image(data3d.astype(dtype), ref_img.affine, header=ref_img.header), path)
    print("Saved:", path)
    return path

# =========================
# First-level: per subject Faces - Houses contrast
# =========================
# Use VT mask (fast). You can swap to haxby.mask[0] for whole-brain.
base_mask_img = nib.load(haxby.mask_vt[0])

contrasts = []
ref_3d = None

print("\nFitting first-level GLMs...")
for idx, (func_path, labels_path) in enumerate(zip(haxby.func, haxby.session_target), start=1):
    func_img = nib.load(func_path)
    if ref_3d is None:
        ref_3d = func_img.slicer[..., 0]  # reference grid

    # resample mask to reference once (consistent grid across subjects)
    mask_img = image.resample_to_img(base_mask_img, ref_3d, interpolation="nearest")
    mask_bool = mask_img.get_fdata().astype(bool)

    labels = np.loadtxt(labels_path, dtype=str, usecols=0)  # first col only
    events = labels_to_events(labels, tr=TR)

    fm = FirstLevelModel(
        t_r=TR,
        mask_img=mask_img,
        hrf_model="spm",
        drift_model="cosine",
        high_pass=1/128,
        noise_model="ar1",
        standardize=False,
        minimize_memory=True,
    )
    fm = fm.fit(func_img, events=events)

    dm = fm.design_matrices_[0]
    cols_lower = [c.lower() for c in dm.columns.tolist()]

    def find_col(prefixes):
        for pref in prefixes:
            for j, c in enumerate(cols_lower):
                if c == pref or c.startswith(pref):
                    return j, dm.columns.tolist()[j]
        return None, None

    j_face, col_face = find_col(["face", "faces"])
    j_house, col_house = find_col(["house", "houses"])
    if j_face is None or j_house is None:
        raise RuntimeError(f"Could not find face/house regressors in subject {idx}. "
                           f"Columns head: {dm.columns.tolist()[:20]}")

    con = np.zeros(len(dm.columns), dtype=float)
    con[j_face] = 1.0
    con[j_house] = -1.0

    con_img = fm.compute_contrast(con, output_type="effect_size")
    con_img = image.resample_to_img(con_img, ref_3d, interpolation="continuous")
    contrasts.append(con_img)

print("Usable subjects:", len(contrasts))

# =========================
# Group analysis: one-sided t/p + your corrections
# =========================
images_4d = np.stack([img.get_fdata().astype(float) for img in contrasts], axis=0)
tmap, pmap = compute_group_t_p(images_4d, mask_bool)
df = images_4d.shape[0] - 1
t0_cluster = float(stats.t.isf(p0_cluster, df=df))

sig_bonf = bonferroni(pmap, mask_bool, alpha)
sig_bh,  pcrit_bh  = stepup_fdr(pmap, mask_bool, q_fdr, mode="BH")
sig_by,  pcrit_by  = stepup_fdr(pmap, mask_bool, q_fdr, mode="BY")
sig_bky, pcrit_bky = stepup_fdr(pmap, mask_bool, q_fdr, mode="BKY")

rng = np.random.default_rng(0)
sig_cluster = cluster_perm_onesample_one_sided_3d(
    images_4d, t0=t0_cluster, alpha=alpha,
    n_perm=n_perm_cluster, mask_bool=mask_bool,
    rng=rng, structure=STRUCT_3D, chunk=chunk
)

print("\nThreshold summary:")
print("  BH  pcrit :", pcrit_bh)
print("  BY  pcrit :", pcrit_by)
print("  BKY pcrit :", pcrit_bky)
print("  Cluster t0:", t0_cluster)

# =========================
# Save outputs
# =========================
tmap_path = save_like(ref_3d, tmap, "group_tmap_faces_minus_houses", np.float32)
pmap_path = save_like(ref_3d, pmap, "group_pmap_faces_minus_houses", np.float32)

save_like(ref_3d, sig_bonf.astype(np.uint8), "sig_Bonf", np.uint8)
save_like(ref_3d, sig_bh.astype(np.uint8),   "sig_BH", np.uint8)
save_like(ref_3d, sig_by.astype(np.uint8),   "sig_BY", np.uint8)
save_like(ref_3d, sig_bky.astype(np.uint8),  "sig_BKY", np.uint8)
save_like(ref_3d, sig_cluster.astype(np.uint8), "sig_ClusterPerm", np.uint8)

# =========================
# Quick evaluation table (real data, no "truth")
# =========================
def summarize(name, sig):
    R = int((sig & mask_bool).sum())
    ncl, mean_sz, med_sz, max_sz = cluster_stats(sig, mask_bool)
    return dict(method=name, R=R, n_clusters=ncl, mean_cluster=mean_sz, median_cluster=med_sz, max_cluster=max_sz)

summary = pd.DataFrame([
    summarize("Bonf", sig_bonf),
    summarize("BH", sig_bh),
    summarize("BY", sig_by),
    summarize("BKY", sig_bky),
    summarize("ClusterPerm", sig_cluster),
])

print("\n=== Haxby real-data method summary ===")
print(summary)
summary.to_csv(os.path.join(OUT_DIR, "haxby_method_summary.csv"), index=False)
print("Saved:", os.path.join(OUT_DIR, "haxby_method_summary.csv"))

# =========================
# Plots
# =========================
plotting.plot_glass_brain(
    nib.load(tmap_path),
    threshold=3.0, colorbar=True,
    title="Haxby: group t-map (Faces - Houses), uncorrected preview t>=3"
)

plotting.plot_glass_brain(
    nib.Nifti1Image(sig_bonf.astype(float), ref_3d.affine),
    threshold=0.5, title="Bonferroni significant voxels"
)
plotting.plot_glass_brain(
    nib.Nifti1Image(sig_bh.astype(float), ref_3d.affine),
    threshold=0.5, title=f"BH significant (q={q_fdr})"
)
plotting.plot_glass_brain(
    nib.Nifti1Image(sig_by.astype(float), ref_3d.affine),
    threshold=0.5, title=f"BY significant (q={q_fdr})"
)
plotting.plot_glass_brain(
    nib.Nifti1Image(sig_bky.astype(float), ref_3d.affine),
    threshold=0.5, title=f"BKY significant (q={q_fdr})"
)
plotting.plot_glass_brain(
    nib.Nifti1Image(sig_cluster.astype(float), ref_3d.affine),
    threshold=0.5, title=f"ClusterPerm (alpha={alpha}, p0={p0_cluster}, perms={n_perm_cluster})"
)

plotting.show()
```

```python
# =========================
# Haxby 3D + your multiple-comparisons methods (Bonf, BH, BY, BKY, ClusterPerm)
# Colab-ready, end-to-end, with outputs for FSLeyes
# =========================

# If needed (uncomment):
# !pip -q install nilearn nibabel scipy pandas scikit-learn matplotlib

import os, shutil, warnings
import numpy as np
import pandas as pd
import nibabel as nib
from scipy import ndimage, stats

from nilearn import datasets, image
from nilearn.glm.first_level import FirstLevelModel

# -------------------------
# User settings
# -------------------------
OUT_DIR = "haxby_final_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# Multiple comparisons settings
alpha = 0.05
q_fdr = 0.10          # q=0.05 can be empty with n=6; 0.10 is still standard for demonstration
p0_cluster = 0.05     # cluster-forming p (one-sided). 0.01 often too strict with n=6
n_perm_cluster = 1000 # increase to 1500/5000 if you want (slower)

# Preprocessing (helps power a lot)
smoothing_fwhm = 6.0  # analogous to your simulation smoothness knob

# Cluster connectivity (26-neighbourhood)
STRUCT_3D = np.ones((3, 3, 3), dtype=int)
_SIGNS = np.array([-1.0, 1.0], dtype=float)
chunk = 64

TR = 2.5  # Haxby TR


# -------------------------
# Helpers
# -------------------------
def labels_to_events(labels_1d, tr):
    import pandas as pd
    labels = np.asarray(labels_1d, dtype=str)
    labels = np.char.lower(np.char.strip(labels))

    onsets, durations, types = [], [], []
    n = labels.size
    start = 0
    while start < n:
        lab = labels[start]
        end = start + 1
        while end < n and labels[end] == lab:
            end += 1
        onsets.append(start * tr)
        durations.append((end - start) * tr)
        types.append(lab)
        start = end
    return pd.DataFrame({"onset": onsets, "duration": durations, "trial_type": types})


def compute_group_t_p(images_4d, mask_bool, eps=1e-8):
    images_4d = np.asarray(images_4d, dtype=float)
    n = images_4d.shape[0]
    df = n - 1

    tmap = np.full(mask_bool.shape, np.nan, dtype=float)
    pmap = np.full(mask_bool.shape, np.nan, dtype=float)

    Xv = images_4d[:, mask_bool]              # (n, V)
    mean = Xv.mean(axis=0)
    sd = Xv.std(axis=0, ddof=1)

    tvals = mean / np.maximum(sd / np.sqrt(n), eps)
    pvals = stats.t.sf(tvals, df=df)          # one-sided

    tmap[mask_bool] = tvals
    pmap[mask_bool] = np.clip(pvals, 1e-300, 1.0)
    return tmap, pmap


def bonferroni(pmap, mask_bool, alpha):
    p = pmap[mask_bool]
    p = p[np.isfinite(p)]
    m = int(p.size)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if m > 0:
        sig[mask_bool] = pmap[mask_bool] <= (alpha / m)
    return sig


def stepup_fdr(pmap, mask_bool, q, mode="BH"):
    p = pmap[mask_bool]
    p = p[np.isfinite(p)]
    m = int(p.size)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if m == 0:
        return sig, np.nan

    p_sorted = np.sort(p)
    i = np.arange(1, m + 1, dtype=float)

    if mode == "BH":
        line = (i / m) * q
        ok = p_sorted <= line
        pcrit = p_sorted[np.max(np.where(ok)[0])] if np.any(ok) else np.nan

    elif mode == "BY":
        c_m = np.sum(1.0 / np.arange(1, m + 1, dtype=float))
        line = (i / m) * (q / c_m)
        ok = p_sorted <= line
        pcrit = p_sorted[np.max(np.where(ok)[0])] if np.any(ok) else np.nan

    elif mode == "BKY":
        q0 = q / (1.0 + q)
        line1 = (i / m) * q0
        ok1 = p_sorted <= line1
        pcrit1 = p_sorted[np.max(np.where(ok1)[0])] if np.any(ok1) else np.nan

        R1 = int(np.sum(p <= pcrit1)) if np.isfinite(pcrit1) else 0
        m0_hat = max(1, m - R1)
        q_star = min(1.0, q0 * (m / m0_hat))

        line2 = (i / m) * q_star
        ok2 = p_sorted <= line2
        pcrit = p_sorted[np.max(np.where(ok2)[0])] if np.any(ok2) else np.nan
    else:
        raise ValueError("mode must be 'BH', 'BY', or 'BKY'")

    if np.isfinite(pcrit):
        sig[mask_bool] = pmap[mask_bool] <= pcrit
    return sig, float(pcrit) if np.isfinite(pcrit) else np.nan


def cluster_perm_onesample_one_sided_3d(images_4d, t0, alpha, n_perm, mask_bool, rng,
                                        structure=STRUCT_3D, eps=1e-12, chunk=64):
    images_4d = np.asarray(images_4d, dtype=float)
    n = images_4d.shape[0]
    df = n - 1

    V = int(mask_bool.sum())
    Xv = images_4d[:, mask_bool]  # (n, V)
    SS = np.sum(Xv**2, axis=0)

    mean_obs = Xv.mean(axis=0)
    var_obs = (SS - n * mean_obs**2) / max(df, 1)
    var_obs = np.maximum(var_obs, 0.0)
    denom_obs = np.sqrt(var_obs / n) + eps
    t_obs_v = mean_obs / denom_obs

    t_obs = np.full(mask_bool.shape, -np.inf, dtype=float)
    t_obs[mask_bool] = t_obs_v

    null_max = np.zeros(n_perm, dtype=int)
    signs_all = rng.choice(_SIGNS, size=(n_perm, n), replace=True)

    for start in range(0, n_perm, chunk):
        stop = min(n_perm, start + chunk)
        signs = signs_all[start:stop]

        mean_perm = (signs @ Xv) / n
        var_perm = (SS[None, :] - n * mean_perm**2) / max(df, 1)
        var_perm = np.maximum(var_perm, 0.0)
        denom_perm = np.sqrt(var_perm / n) + eps
        t_perm = mean_perm / denom_perm

        for b in range(stop - start):
            tmap_b = np.full(mask_bool.shape, -np.inf, dtype=float)
            tmap_b[mask_bool] = t_perm[b]
            supra = (tmap_b >= t0) & mask_bool
            lab, nlab = ndimage.label(supra, structure=structure)
            if nlab == 0:
                null_max[start + b] = 0
            else:
                counts = np.bincount(lab.ravel())
                null_max[start + b] = int(counts[1:].max(initial=0))

    supra_obs = (t_obs >= t0) & mask_bool
    lab, nlab = ndimage.label(supra_obs, structure=structure)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if nlab == 0:
        return sig

    sizes_obs = np.bincount(lab.ravel())[1:]
    ge = (null_max[:, None] >= sizes_obs[None, :])
    pvals = (1.0 + ge.sum(axis=0)) / (n_perm + 1.0)

    for k, p in enumerate(pvals, start=1):
        if p <= alpha:
            sig[lab == k] = True
    return sig


def cluster_stats(sig_mask, mask_bool, structure=STRUCT_3D):
    sig = sig_mask.astype(bool) & mask_bool
    lab, nlab = ndimage.label(sig, structure=structure)
    if nlab == 0:
        return 0, np.nan, np.nan, 0
    sizes = np.bincount(lab.ravel())[1:]
    return int(nlab), float(np.mean(sizes)), float(np.median(sizes)), int(np.max(sizes))


def save_like(ref_img, data3d, name, dtype=np.float32):
    path = os.path.join(OUT_DIR, f"{name}.nii.gz")
    nib.save(nib.Nifti1Image(data3d.astype(dtype), ref_img.affine, header=ref_img.header), path)
    print("Saved:", path)
    return path


def apply_mask_to_t(tmap, sig, mask_bool):
    out = np.zeros_like(tmap, dtype=np.float32)
    keep = sig & mask_bool
    out[keep] = tmap[keep]
    return out


# -------------------------
# Fetch Haxby subjects 1..6 (fresh dir to avoid partial cache)
# -------------------------
DATA_DIR = "/content/nilearn_haxby_full"
if os.path.exists(DATA_DIR):
    shutil.rmtree(DATA_DIR)

print("Fetching Haxby subjects 1..6 into:", DATA_DIR)
haxby = datasets.fetch_haxby(data_dir=DATA_DIR, subjects=[1, 2, 3, 4, 5, 6])
print("Subjects fetched:", len(haxby.func))

if len(haxby.func) < 3:
    raise RuntimeError("Need >=3 subjects for group inference and cluster permutation.")

# Use VT mask (faster, more interpretable). Change to haxby.mask[0] if you want whole brain.
base_mask_img = nib.load(haxby.mask_vt[0])

# Reference grid: subj1 first volume
ref_func_img = nib.load(haxby.func[0])
ref_3d = ref_func_img.slicer[..., 0]

# Resample mask ONCE to reference grid (consistent across subjects)
with warnings.catch_warnings():
    warnings.simplefilter("ignore")  # suppress sform warnings
    mask_img = image.resample_to_img(base_mask_img, ref_3d, interpolation="nearest")
mask_bool = mask_img.get_fdata().astype(bool)

# -------------------------
# First-level GLM per subject -> Faces - Houses subject contrast (effect_size)
# -------------------------
contrasts = []
print("\nFitting first-level GLMs (smoothing_fwhm=", smoothing_fwhm, ") ...")

for subj_idx, (func_path, labels_path) in enumerate(zip(haxby.func, haxby.session_target), start=1):
    func_img = nib.load(func_path)

    labels = np.loadtxt(labels_path, dtype=str, usecols=0)
    events = labels_to_events(labels, tr=TR)

    fm = FirstLevelModel(
        t_r=TR,
        mask_img=mask_img,
        hrf_model="spm",
        drift_model="cosine",
        high_pass=1/128,
        noise_model="ar1",
        smoothing_fwhm=smoothing_fwhm,
        standardize=False,
        minimize_memory=True,
    ).fit(func_img, events=events)

    dm = fm.design_matrices_[0]
    cols_lower = [c.lower() for c in dm.columns.tolist()]

    def find_col(prefixes):
        for pref in prefixes:
            for j, c in enumerate(cols_lower):
                if c == pref or c.startswith(pref):
                    return j
        return None

    j_face = find_col(["face", "faces"])
    j_house = find_col(["house", "houses"])
    if j_face is None or j_house is None:
        raise RuntimeError(f"Could not find face/house regressors for subject {subj_idx}. "
                           f"Design cols head: {dm.columns.tolist()[:20]}")

    con = np.zeros(len(dm.columns), dtype=float)
    con[j_face] = 1.0
    con[j_house] = -1.0

    con_img = fm.compute_contrast(con, output_type="effect_size")
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        con_img = image.resample_to_img(con_img, ref_3d, interpolation="continuous")

    contrasts.append(con_img)

print("Usable subjects:", len(contrasts))

# -------------------------
# Group one-sample t/p (one-sided)
# -------------------------
images_4d = np.stack([img.get_fdata().astype(float) for img in contrasts], axis=0)
tmap, pmap = compute_group_t_p(images_4d, mask_bool)

df = images_4d.shape[0] - 1
t0_cluster = float(stats.t.isf(p0_cluster, df=df))
print("\nGroup df:", df, "cluster-forming t0:", t0_cluster)

# -------------------------
# Apply your corrections
# -------------------------
sig_bonf = bonferroni(pmap, mask_bool, alpha)
sig_bh,  pcrit_bh  = stepup_fdr(pmap, mask_bool, q_fdr, mode="BH")
sig_by,  pcrit_by  = stepup_fdr(pmap, mask_bool, q_fdr, mode="BY")
sig_bky, pcrit_bky = stepup_fdr(pmap, mask_bool, q_fdr, mode="BKY")

rng = np.random.default_rng(0)
sig_cluster = cluster_perm_onesample_one_sided_3d(
    images_4d, t0=t0_cluster, alpha=alpha, n_perm=n_perm_cluster,
    mask_bool=mask_bool, rng=rng, structure=STRUCT_3D, chunk=chunk
)

print("\nThreshold summary:")
print("  BH  pcrit :", pcrit_bh)
print("  BY  pcrit :", pcrit_by)
print("  BKY pcrit :", pcrit_bky)

# -------------------------
# Save outputs for FSLeyes
# -------------------------
save_like(ref_3d, tmap, "group_tmap_faces_minus_houses", np.float32)
save_like(ref_3d, pmap, "group_pmap_faces_minus_houses", np.float32)
save_like(ref_3d, mask_bool.astype(np.uint8), "analysis_mask_vt", np.uint8)

save_like(ref_3d, sig_bonf.astype(np.uint8), "sig_Bonf", np.uint8)
save_like(ref_3d, sig_bh.astype(np.uint8),   "sig_BH", np.uint8)
save_like(ref_3d, sig_by.astype(np.uint8),   "sig_BY", np.uint8)
save_like(ref_3d, sig_bky.astype(np.uint8),  "sig_BKY", np.uint8)
save_like(ref_3d, sig_cluster.astype(np.uint8), "sig_ClusterPerm", np.uint8)

# Thresholded t-maps (look nicer in FSLeyes than binary masks)
save_like(ref_3d, apply_mask_to_t(tmap, sig_bonf, mask_bool), "tmap_sig_Bonf", np.float32)
save_like(ref_3d, apply_mask_to_t(tmap, sig_bh,   mask_bool), "tmap_sig_BH", np.float32)
save_like(ref_3d, apply_mask_to_t(tmap, sig_by,   mask_bool), "tmap_sig_BY", np.float32)
save_like(ref_3d, apply_mask_to_t(tmap, sig_bky,  mask_bool), "tmap_sig_BKY", np.float32)
save_like(ref_3d, apply_mask_to_t(tmap, sig_cluster, mask_bool), "tmap_sig_ClusterPerm", np.float32)

# -------------------------
# Real-data evaluation table (no ground truth): R + cluster sizes
# -------------------------
def summarize(name, sig):
    R = int((sig & mask_bool).sum())
    ncl, mean_sz, med_sz, max_sz = cluster_stats(sig, mask_bool)
    return dict(method=name, R=R, n_clusters=ncl, mean_cluster=mean_sz, median_cluster=med_sz, max_cluster=max_sz)

summary = pd.DataFrame([
    summarize("Bonf", sig_bonf),
    summarize("BH", sig_bh),
    summarize("BY", sig_by),
    summarize("BKY", sig_bky),
    summarize("ClusterPerm", sig_cluster),
])

print("\n=== Haxby real-data method summary ===")
print(summary)
summary.to_csv(os.path.join(OUT_DIR, "haxby_method_summary.csv"), index=False)
print("Saved:", os.path.join(OUT_DIR, "haxby_method_summary.csv"))

# -------------------------
# Zip outputs for download (optional)
# -------------------------
zip_path = f"{OUT_DIR}.zip"
!zip -qr {zip_path} {OUT_DIR}
print("\nDownload this zip from Colab file browser:", zip_path)
print("Then open in FSLeyes with e.g.:")
print(f"  fsleyes {OUT_DIR}/group_tmap_faces_minus_houses.nii.gz "
      f"{OUT_DIR}/tmap_sig_BH.nii.gz {OUT_DIR}/tmap_sig_BY.nii.gz "
      f"{OUT_DIR}/tmap_sig_BKY.nii.gz {OUT_DIR}/tmap_sig_ClusterPerm.nii.gz")
```

```python
# =========================
# FINAL: Haxby-only, 3D, your methods (Bonf, BH, BY, BKY, ClusterPerm)
# Runs BOTH one-sided directions (POS and NEG) + optional two-sided p-map
# Saves NIfTIs + thresholded t-overlays for FSLeyes + summary CSVs
# Colab-ready
# =========================

# If needed:
# !pip -q install nilearn nibabel scipy pandas scikit-learn matplotlib

import os, shutil, warnings
import numpy as np
import pandas as pd
import nibabel as nib
from scipy import ndimage, stats

from nilearn import datasets, image
from nilearn.glm.first_level import FirstLevelModel
from nilearn.masking import compute_epi_mask

# -------------------------
# Settings
# -------------------------
OUT_DIR = "haxby_final_outputs_diraware"
os.makedirs(OUT_DIR, exist_ok=True)

# Control levels
alpha = 0.05
q_fdr = 0.10

# Cluster permutation
p0_cluster = 0.05       # cluster-forming p (one-sided)
n_perm_cluster = 1000   # increase later (1500/5000) if desired
chunk = 64
STRUCT_3D = np.ones((3, 3, 3), dtype=int)
_SIGNS = np.array([-1.0, 1.0], dtype=float)

# First-level preprocessing
TR = 2.5
smoothing_fwhm = 6.0

# Subjects (Haxby has 6)
SUBJECTS = [1, 2, 3, 4, 5, 6]

# Fresh download dir (avoids partial cache)
DATA_DIR = "/content/nilearn_haxby_full"
if os.path.exists(DATA_DIR):
    shutil.rmtree(DATA_DIR)


# -------------------------
# Helpers
# -------------------------
def labels_to_events(labels_1d, tr):
    labels = np.asarray(labels_1d, dtype=str)
    labels = np.char.lower(np.char.strip(labels))
    onsets, durations, types = [], [], []
    n = labels.size
    start = 0
    while start < n:
        lab = labels[start]
        end = start + 1
        while end < n and labels[end] == lab:
            end += 1
        onsets.append(start * tr)
        durations.append((end - start) * tr)
        types.append(lab)
        start = end
    return pd.DataFrame({"onset": onsets, "duration": durations, "trial_type": types})


def compute_group_t_p(images_4d, mask_bool, one_sided="pos", eps=1e-8):
    """
    images_4d: (n, X, Y, Z) subject contrast values (effect-size)
    one_sided: "pos" -> P(T >= t), "neg" -> P(T <= t), "two" -> 2*P(T >= |t|)
    """
    images_4d = np.asarray(images_4d, dtype=float)
    n = images_4d.shape[0]
    df = n - 1

    tmap = np.full(mask_bool.shape, np.nan, dtype=float)
    pmap = np.full(mask_bool.shape, np.nan, dtype=float)

    Xv = images_4d[:, mask_bool]
    mean = Xv.mean(axis=0)
    sd = Xv.std(axis=0, ddof=1)
    tvals = mean / np.maximum(sd / np.sqrt(n), eps)
    tmap[mask_bool] = tvals

    if one_sided == "pos":
        pvals = stats.t.sf(tvals, df=df)                 # P(T >= t)
    elif one_sided == "neg":
        pvals = stats.t.cdf(tvals, df=df)                # P(T <= t)
    elif one_sided == "two":
        pvals = 2.0 * stats.t.sf(np.abs(tvals), df=df)   # two-sided
    else:
        raise ValueError("one_sided must be 'pos', 'neg', or 'two'")

    pmap[mask_bool] = np.clip(pvals, 1e-300, 1.0)
    return tmap, pmap, df


def bonferroni(pmap, mask_bool, alpha):
    p = pmap[mask_bool]
    p = p[np.isfinite(p)]
    m = int(p.size)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if m > 0:
        sig[mask_bool] = pmap[mask_bool] <= (alpha / m)
    return sig


def stepup_fdr(pmap, mask_bool, q, mode="BH"):
    p = pmap[mask_bool]
    p = p[np.isfinite(p)]
    m = int(p.size)
    sig = np.zeros_like(mask_bool, dtype=bool)
    if m == 0:
        return sig, np.nan

    p_sorted = np.sort(p)
    i = np.arange(1, m + 1, dtype=float)

    if mode == "BH":
        line = (i / m) * q
        ok = p_sorted <= line
        pcrit = p_sorted[np.max(np.where(ok)[0])] if np.any(ok) else np.nan

    elif mode == "BY":
        c_m = np.sum(1.0 / np.arange(1, m + 1, dtype=float))
        line = (i / m) * (q / c_m)
        ok = p_sorted <= line
        pcrit = p_sorted[np.max(np.where(ok)[0])] if np.any(ok) else np.nan

    elif mode == "BKY":
        q0 = q / (1.0 + q)
        line1 = (i / m) * q0
        ok1 = p_sorted <= line1
        pcrit1 = p_sorted[np.max(np.where(ok1)[0])] if np.any(ok1) else np.nan
        R1 = int(np.sum(p <= pcrit1)) if np.isfinite(pcrit1) else 0
        m0_hat = max(1, m - R1)
        q_star = min(1.0, q0 * (m / m0_hat))
        line2 = (i / m) * q_star
        ok2 = p_sorted <= line2
        pcrit = p_sorted[np.max(np.where(ok2)[0])] if np.any(ok2) else np.nan
    else:
        raise ValueError("mode must be 'BH', 'BY', or 'BKY'")

    if np.isfinite(pcrit):
        sig[mask_bool] = pmap[mask_bool] <= pcrit
    return sig, float(pcrit) if np.isfinite(pcrit) else np.nan


def cluster_perm_signflip_onesided(images_4d, t0, alpha, n_perm, mask_bool, rng,
                                  structure=STRUCT_3D, direction="pos", eps=1e-12, chunk=64):
    """
    direction="pos" tests large positive clusters: supra = t >= t0
    direction="neg" tests large negative clusters: supra = t <= -t0
    """
    images_4d = np.asarray(images_4d, dtype=float)
    n = images_4d.shape[0]
    df = n - 1

    Xv = images_4d[:, mask_bool]  # (n, V)
    SS = np.sum(Xv**2, axis=0)

    # observed t
    mean_obs = Xv.mean(axis=0)
    var_obs = (SS - n * mean_obs**2) / max(df, 1)
    var_obs = np.maximum(var_obs, 0.0)
    denom_obs = np.sqrt(var_obs / n) + eps
    t_obs_v = mean_obs / denom_obs

    t_obs = np.full(mask_bool.shape, np.nan, dtype=float)
    t_obs[mask_bool] = t_obs_v

    if direction == "pos":
        supra_obs = (t_obs >= t0) & mask_bool
    elif direction == "neg":
        supra_obs = (t_obs <= -t0) & mask_bool
    else:
        raise ValueError("direction must be 'pos' or 'neg'")

    lab, nlab = ndimage.label(supra_obs, structure=structure)
    if nlab == 0:
        return np.zeros_like(mask_bool, dtype=bool)

    sizes_obs = np.bincount(lab.ravel())[1:]

    # null distribution of max cluster size
    null_max = np.zeros(n_perm, dtype=int)
    signs_all = rng.choice(_SIGNS, size=(n_perm, n), replace=True)

    for start in range(0, n_perm, chunk):
        stop = min(n_perm, start + chunk)
        signs = signs_all[start:stop]

        mean_perm = (signs @ Xv) / n
        var_perm = (SS[None, :] - n * mean_perm**2) / max(df, 1)
        var_perm = np.maximum(var_perm, 0.0)
        denom_perm = np.sqrt(var_perm / n) + eps
        t_perm = mean_perm / denom_perm

        for b in range(stop - start):
            tmap_b = np.full(mask_bool.shape, np.nan, dtype=float)
            tmap_b[mask_bool] = t_perm[b]

            if direction == "pos":
                supra = (tmap_b >= t0) & mask_bool
            else:
                supra = (tmap_b <= -t0) & mask_bool

            lab_b, nlab_b = ndimage.label(supra, structure=structure)
            if nlab_b == 0:
                null_max[start + b] = 0
            else:
                null_max[start + b] = int(np.bincount(lab_b.ravel())[1:].max(initial=0))

    # cluster p-values
    ge = (null_max[:, None] >= sizes_obs[None, :])
    pvals = (1.0 + ge.sum(axis=0)) / (n_perm + 1.0)

    sig = np.zeros_like(mask_bool, dtype=bool)
    for k, p in enumerate(pvals, start=1):
        if p <= alpha:
            sig[lab == k] = True
    return sig


def cluster_stats(sig_mask, mask_bool, structure=STRUCT_3D):
    sig = sig_mask.astype(bool) & mask_bool
    lab, nlab = ndimage.label(sig, structure=structure)
    if nlab == 0:
        return 0, np.nan, np.nan, 0
    sizes = np.bincount(lab.ravel())[1:]
    return int(nlab), float(np.mean(sizes)), float(np.median(sizes)), int(np.max(sizes))


def apply_mask_to_t(tmap, sig, mask_bool):
    out = np.zeros_like(tmap, dtype=np.float32)
    keep = sig & mask_bool
    out[keep] = tmap[keep]
    return out


def save_like(ref_img, data3d, name, dtype=np.float32):
    path = os.path.join(OUT_DIR, f"{name}.nii.gz")
    nib.save(nib.Nifti1Image(data3d.astype(dtype), ref_img.affine, header=ref_img.header), path)
    print("Saved:", path)
    return path


def summarize_methods(prefix, tmap, pmap, mask_bool, sigs_dict):
    rows = []
    pm = pmap[mask_bool]
    rows.append({"method": "uncorrected_p<0.05", "R": int(np.sum(pm < 0.05))})
    for name, sig in sigs_dict.items():
        R = int(np.sum(sig & mask_bool))
        ncl, mean_sz, med_sz, max_sz = cluster_stats(sig, mask_bool)
        rows.append({
            "method": name,
            "R": R,
            "n_clusters": ncl,
            "mean_cluster": mean_sz,
            "median_cluster": med_sz,
            "max_cluster": max_sz
        })
    df = pd.DataFrame(rows)
    df.to_csv(os.path.join(OUT_DIR, f"summary_{prefix}.csv"), index=False)
    return df


# -------------------------
# Fetch Haxby (6 subjects)
# -------------------------
print("Fetching Haxby subjects", SUBJECTS, "into:", DATA_DIR)
haxby = datasets.fetch_haxby(data_dir=DATA_DIR, subjects=SUBJECTS)
print("Subjects fetched:", len(haxby.func))
if len(haxby.func) < 3:
    raise RuntimeError("Need >=3 subjects.")

# Reference image/grid
ref_func = nib.load(haxby.func[0])
ref_3d = ref_func.slicer[..., 0]

# EPI-derived mask (aligned by construction)
print("Computing EPI mask from subject 1...")
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    epi_mask_img = compute_epi_mask(ref_func)
    epi_mask_img = image.resample_to_img(epi_mask_img, ref_3d, interpolation="nearest")
mask_bool = epi_mask_img.get_fdata().astype(bool)
print("Mask voxels:", int(mask_bool.sum()))
save_like(ref_3d, mask_bool.astype(np.uint8), "analysis_mask_epi", np.uint8)

# -------------------------
# First-level GLM per subject -> effect-size contrast Faces - Houses
# -------------------------
print("\nFitting first-level GLMs (smoothing_fwhm=", smoothing_fwhm, ") ...")
contrasts = []

for subj_idx, (func_path, labels_path) in enumerate(zip(haxby.func, haxby.session_target), start=1):
    func_img = nib.load(func_path)

    labels = np.loadtxt(labels_path, dtype=str, usecols=0)
    events = labels_to_events(labels, tr=TR)

    fm = FirstLevelModel(
        t_r=TR,
        mask_img=epi_mask_img,
        hrf_model="spm",
        drift_model="cosine",
        high_pass=1/128,
        noise_model="ar1",
        smoothing_fwhm=smoothing_fwhm,
        standardize=False,
        minimize_memory=True,
    ).fit(func_img, events=events)

    dm = fm.design_matrices_[0]
    cols_lower = [c.lower() for c in dm.columns.tolist()]

    def find_col(prefixes):
        for pref in prefixes:
            for j, c in enumerate(cols_lower):
                if c == pref or c.startswith(pref):
                    return j
        return None

    j_face = find_col(["face", "faces"])
    j_house = find_col(["house", "houses"])
    if j_face is None or j_house is None:
        raise RuntimeError(f"Face/House regressors not found for subject {subj_idx}. "
                           f"Design cols head: {dm.columns.tolist()[:20]}")

    con = np.zeros(len(dm.columns), dtype=float)
    con[j_face] = 1.0
    con[j_house] = -1.0

    con_img = fm.compute_contrast(con, output_type="effect_size")
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        con_img = image.resample_to_img(con_img, ref_3d, interpolation="continuous")
    contrasts.append(con_img)

print("Usable subjects:", len(contrasts))

images_4d = np.stack([img.get_fdata().astype(float) for img in contrasts], axis=0)
n_subj = images_4d.shape[0]
df = n_subj - 1

# Cluster-forming t0
t0_cluster = float(stats.t.isf(p0_cluster, df=df))
print("\nGroup df:", df, "| cluster-forming p0:", p0_cluster, "| t0:", t0_cluster)

# -------------------------
# Run both directions
# -------------------------
results = {}

for direction in ["pos", "neg"]:
    print("\n=========================")
    print("Running direction:", direction.upper())
    print("=========================")

    # voxelwise pmap in chosen direction
    tmap, pmap, _df = compute_group_t_p(images_4d, mask_bool, one_sided=direction)

    # corrections (your methods)
    sig_bonf = bonferroni(pmap, mask_bool, alpha)
    sig_bh,  pcrit_bh  = stepup_fdr(pmap, mask_bool, q_fdr, mode="BH")
    sig_by,  pcrit_by  = stepup_fdr(pmap, mask_bool, q_fdr, mode="BY")
    sig_bky, pcrit_bky = stepup_fdr(pmap, mask_bool, q_fdr, mode="BKY")

    rng = np.random.default_rng(0 if direction == "pos" else 1)
    sig_cluster = cluster_perm_signflip_onesided(
        images_4d, t0=t0_cluster, alpha=alpha,
        n_perm=n_perm_cluster, mask_bool=mask_bool, rng=rng,
        structure=STRUCT_3D, direction=direction, chunk=chunk
    )

    print("Threshold summary (pcrit): BH=", pcrit_bh, " BY=", pcrit_by, " BKY=", pcrit_bky)

    prefix = f"{direction}"
    # save base maps
    save_like(ref_3d, tmap, f"{prefix}_group_tmap", np.float32)
    save_like(ref_3d, pmap, f"{prefix}_group_pmap", np.float32)

    # save sig masks
    save_like(ref_3d, sig_bonf.astype(np.uint8),   f"{prefix}_sig_Bonf", np.uint8)
    save_like(ref_3d, sig_bh.astype(np.uint8),     f"{prefix}_sig_BH", np.uint8)
    save_like(ref_3d, sig_by.astype(np.uint8),     f"{prefix}_sig_BY", np.uint8)
    save_like(ref_3d, sig_bky.astype(np.uint8),    f"{prefix}_sig_BKY", np.uint8)
    save_like(ref_3d, sig_cluster.astype(np.uint8),f"{prefix}_sig_ClusterPerm", np.uint8)

    # save thresholded t overlays (best for FSLeyes)
    save_like(ref_3d, apply_mask_to_t(tmap, sig_bh, mask_bool),     f"{prefix}_tmap_sig_BH", np.float32)
    save_like(ref_3d, apply_mask_to_t(tmap, sig_by, mask_bool),     f"{prefix}_tmap_sig_BY", np.float32)
    save_like(ref_3d, apply_mask_to_t(tmap, sig_bky, mask_bool),    f"{prefix}_tmap_sig_BKY", np.float32)
    save_like(ref_3d, apply_mask_to_t(tmap, sig_cluster, mask_bool),f"{prefix}_tmap_sig_ClusterPerm", np.float32)

    # diagnostics
    pm = pmap[mask_bool]
    tm = tmap[mask_bool]
    print("Diagnostics in mask:")
    print("  t min/mean/max:", float(np.nanmin(tm)), float(np.nanmean(tm)), float(np.nanmax(tm)))
    print("  p min/mean/max:", float(np.nanmin(pm)), float(np.nanmean(pm)), float(np.nanmax(pm)))
    print("  uncorrected p<0.05 voxels:", int(np.sum(pm < 0.05)))

    sigs = {
        "Bonf": sig_bonf,
        "BH": sig_bh,
        "BY": sig_by,
        "BKY": sig_bky,
        "ClusterPerm": sig_cluster
    }
    summary_df = summarize_methods(prefix, tmap, pmap, mask_bool, sigs)
    print("\nSummary table:", direction.upper())
    print(summary_df)

    results[direction] = {
        "tmap": tmap, "pmap": pmap,
        "sigs": sigs,
        "pcrit": {"BH": pcrit_bh, "BY": pcrit_by, "BKY": pcrit_bky},
        "summary": summary_df
    }

# Optional: two-sided p-map (voxelwise only; not running cluster permutation here)
print("\n=========================")
print("Two-sided voxelwise p-map (optional)")
print("=========================")
tmap_two, pmap_two, _ = compute_group_t_p(images_4d, mask_bool, one_sided="two")
save_like(ref_3d, tmap_two, "two_group_tmap", np.float32)
save_like(ref_3d, pmap_two, "two_group_pmap", np.float32)

# Zip outputs for download
zip_path = f"{OUT_DIR}.zip"
!zip -qr {zip_path} {OUT_DIR}
print("\nDownload:", zip_path)
print("\nFSLeyes examples:")
print(f"  fsleyes {OUT_DIR}/pos_group_tmap.nii.gz {OUT_DIR}/pos_tmap_sig_BH.nii.gz {OUT_DIR}/pos_tmap_sig_ClusterPerm.nii.gz")
print(f"  fsleyes {OUT_DIR}/neg_group_tmap.nii.gz {OUT_DIR}/neg_tmap_sig_BH.nii.gz {OUT_DIR}/neg_tmap_sig_ClusterPerm.nii.gz")
```

```python
import numpy as np
from nilearn.datasets import fetch_haxby
from nilearn.image import load_img, index_img
from nilearn.masking import apply_mask, unmask
from scipy import stats
from scipy import ndimage
from nilearn.image import mean_img
from nilearn.image import math_img
from nilearn import plotting
from nilearn.image import new_img_like
# -----------------------------
# 1) Pull Haxby data
# -----------------------------
haxby = fetch_haxby(data_dir="./nilearn_data")

func_path = haxby.func[0]                 # 4D nifti
labels_path = haxby.session_target[0]     # text labels per TR
mask_path = haxby.mask                    # 3D mask nifti

labels = np.loadtxt(labels_path, dtype=str, usecols=0)

# Load 4D func
func_img = load_img(func_path)

# -----------------------------
# 2) Build a per-run "effect map"
#    Here: mean(Face) - mean(House)
# -----------------------------
# Find timepoints (TR indices) for the two conditions
face_idx = np.where(labels == "face")[0]
house_idx = np.where(labels == "house")[0]

if face_idx.size == 0 or house_idx.size == 0:
    raise ValueError("Could not find 'face' or 'house' in session_target labels.")

# Compute mean image for each condition
face_mean_img  = mean_img(index_img(func_img, face_idx))
house_mean_img = mean_img(index_img(func_img, house_idx))

# Contrast image (faces - houses)
contrast_img = math_img("img1 - img2", img1=face_mean_img, img2=house_mean_img)

# Vectorize within mask
contrast_vec = apply_mask(contrast_img, mask_path)  # shape (n_vox_in_mask,)

# -----------------------------
# 3) Create a stack of "images" like your sim code expects
#    Your sim expects multiple subjects. With Haxby subject-2 only,
#    we treat each RUN as one sample IF you load multiple runs.
#
#    fetch_haxby() gives one run by default; easiest is to do
#    a split-half trick to create multiple samples:
# -----------------------------
# Split timepoints into 4 chunks -> 4 pseudo-runs (simple + fast)
T = func_img.shape[-1]
chunks = np.array_split(np.arange(T), 4)

samples = []
for idx in chunks:
    lab_chunk = labels[idx]
    face = idx[lab_chunk == "face"]
    house = idx[lab_chunk == "house"]
    if face.size == 0 or house.size == 0:
        continue
    face_mean  = mean_img(index_img(func_img, face))
    house_mean = mean_img(index_img(func_img, house))
    con = math_img("img1 - img2", img1=face_mean, img2=house_mean)
    samples.append(apply_mask(con, mask_path))
images = np.vstack(samples)  # shape (n_samples, n_vox)
n = images.shape[0]
print("n_samples:", n, "n_vox:", images.shape[1])

# -----------------------------
# 4) One-sample voxelwise t-test across samples
#    (same logic as your compute_t_p)
# -----------------------------
mean = images.mean(axis=0)
sd = images.std(axis=0, ddof=1)
eps = 1e-8
print("n samples:", images.shape[0])

tmap = mean / np.maximum(sd / np.sqrt(n), eps)

# one-sided p-values (activation only)
pmap = stats.t.sf(tmap, df=n - 1)
pmap = np.clip(pmap, 1e-300, 1.0)

# -----------------------------
# 5) Apply your multiple comparisons corrections
# -----------------------------
alpha = 0.05
q_fdr = 0.05
m = pmap.size

# Bonferroni
bonf = pmap <= (alpha / m)

# BH
p_sorted = np.sort(pmap)
bh_line = (np.arange(1, m + 1) / m) * q_fdr
bh_ok = p_sorted <= bh_line
pcrit_bh = p_sorted[np.max(np.where(bh_ok)[0])] if np.any(bh_ok) else np.nan
bh = pmap <= pcrit_bh if np.isfinite(pcrit_bh) else np.zeros_like(pmap, bool)

# BY
c_m = np.sum(1.0 / np.arange(1, m + 1))
by_line = (np.arange(1, m + 1) / m) * (q_fdr / c_m)
by_ok = p_sorted <= by_line
pcrit_by = p_sorted[np.max(np.where(by_ok)[0])] if np.any(by_ok) else np.nan
by = pmap <= pcrit_by if np.isfinite(pcrit_by) else np.zeros_like(pmap, bool)

# BKY (two-stage)
q0 = q_fdr / (1.0 + q_fdr)
bky1_line = (np.arange(1, m + 1) / m) * q0
bky1_ok = p_sorted <= bky1_line
pcrit1 = p_sorted[np.max(np.where(bky1_ok)[0])] if np.any(bky1_ok) else np.nan
R1 = int(np.sum(pmap <= pcrit1)) if np.isfinite(pcrit1) else 0
m0_hat = max(1, m - R1)
q_star = min(1.0, q0 * (m / m0_hat))

bky2_line = (np.arange(1, m + 1) / m) * q_star
bky2_ok = p_sorted <= bky2_line
pcrit_bky = p_sorted[np.max(np.where(bky2_ok)[0])] if np.any(bky2_ok) else np.nan
bky = pmap <= pcrit_bky if np.isfinite(pcrit_bky) else np.zeros_like(pmap, bool)

print("Rejections:")
print(" Bonf:", int(bonf.sum()))
print(" BH  :", int(bh.sum()))
print(" BY  :", int(by.sum()))
print(" BKY :", int(bky.sum()))

# -----------------------------
# 6) Cluster permutation (one-sided sign-flip) on the *maps*
#    We do sign-flips at the sample-map level (n samples)
# -----------------------------
p0_cluster = 0.001
n_perm = 1000
t0 = float(stats.t.isf(p0_cluster, df=n - 1))

# cluster connectivity in 3D needs a 3x3x3 struct; we operate in voxel-space
# easiest: convert masks to 3D images and use nilearn's mask to unmask
# We'll do permutation in voxel-vector space, then reshape via unmask for labeling.
mask_bool_img = load_img(mask_path)
mask_data = mask_bool_img.get_fdata().astype(bool)
struct3 = np.ones((3, 3, 3), dtype=int)

def max_cluster_extent_3d(tvals_vec):
    t_img = unmask(tvals_vec, mask_path).get_fdata()
    supra = (t_img >= t0) & mask_data
    lab, nlab = ndimage.label(supra, structure=struct3)
    if nlab == 0:
        return 0, lab
    # cluster sizes
    sizes = ndimage.sum(np.ones_like(t_img, int), labels=lab, index=np.arange(1, nlab + 1))
    return int(np.max(sizes)), lab

# observed
t_obs_vec = tmap
_, lab_obs = max_cluster_extent_3d(t_obs_vec)

# perm null
rng = np.random.default_rng(0)
null_max = np.empty(n_perm, dtype=int)
for b in range(n_perm):
    signs = rng.choice([-1.0, 1.0], size=(n, 1))
    imgp = images * signs
    meanp = imgp.mean(axis=0)
    sdp = imgp.std(axis=0, ddof=1)
    tp = meanp / np.maximum(sdp / np.sqrt(n), eps)
    null_max[b], _ = max_cluster_extent_3d(tp)

# clusterwise p-values for observed clusters
sig_cluster = np.zeros(mask_data.shape, dtype=bool)
t_img_obs = unmask(t_obs_vec, mask_path).get_fdata()
supra_obs = (t_img_obs >= t0) & mask_data
lab, nlab = ndimage.label(supra_obs, structure=struct3)

if nlab > 0:
    sizes = ndimage.sum(np.ones_like(t_img_obs, int), labels=lab, index=np.arange(1, nlab + 1)).astype(int)
    for k, s in enumerate(sizes, start=1):
        p_cl = (1.0 + np.sum(null_max >= s)) / (n_perm + 1.0)
        if p_cl <= alpha:
            sig_cluster[lab == k] = True

print(" ClusterPerm clusters:", int(nlab), "sig vox:", int(sig_cluster.sum()))


plotting.plot_stat_map(
    t_map,
    bg_img=anat_img,
    threshold=0,
    title="t-map (faces > houses)"
)

plotting.plot_roi(
    bh_mask,
    bg_img=anat_img,
    title="BH significant voxels (q=0.05)"
)

plotting.show()

# -----------------------------
# 7) Visualize one method on the brain
# -----------------------------

def show_mask(vec_bool, title):
    img = unmask(vec_bool.astype(int), mask_path)  # 0/1 image
    plotting.plot_roi(img, bg_img=haxby.anat[0], title=title)
    plotting.show()

show_mask(bh, "BH significant voxels (faces > houses)")
```

```python
import numpy as np
import nibabel as nib
import pandas as pd
from nilearn.datasets import fetch_haxby
from nilearn.image import load_img, index_img, mean_img, math_img, resample_to_img
from nilearn.masking import apply_mask, unmask, compute_epi_mask
from nilearn import plotting
from scipy import stats, ndimage

# -----------------------------
# 1) Pull Haxby data (ONE subject)
# -----------------------------
haxby = fetch_haxby(data_dir="./nilearn_data")  # default is a single subject

func_path = haxby.func[0]
labels_path = haxby.session_target[0]
anat_path = haxby.anat[0]

labels = np.loadtxt(labels_path, dtype=str, usecols=0)
labels = np.char.lower(np.char.strip(labels))

func_img = load_img(func_path)
anat_img = load_img(anat_path)

# -----------------------------
# 2) Use an EPI-derived mask (guaranteed aligned)
# -----------------------------
mask_img = compute_epi_mask(func_img)  # aligned with func_img by construction

# -----------------------------
# 3) Build pseudo-samples: split timepoints into chunks
# -----------------------------
T = func_img.shape[-1]
n_chunks = 6  # more chunks -> more pseudo-samples (still not independent)
chunks = np.array_split(np.arange(T), n_chunks)

samples = []
kept = 0
for idx in chunks:
    lab_chunk = labels[idx]
    face = idx[lab_chunk == "face"]
    house = idx[lab_chunk == "house"]
    if face.size == 0 or house.size == 0:
        continue

    face_mean  = mean_img(index_img(func_img, face))
    house_mean = mean_img(index_img(func_img, house))
    con_img = math_img("img1 - img2", img1=face_mean, img2=house_mean)

    # vectorize within mask
    samples.append(apply_mask(con_img, mask_img))
    kept += 1

images = np.vstack(samples)  # (n_samples, n_vox)
n = images.shape[0]
m = images.shape[1]
print("n_samples:", n, "n_vox:", m)
if n < 3:
    raise RuntimeError("Too few pseudo-samples; increase n_chunks or check labels.")

# -----------------------------
# 4) One-sample t-test across pseudo-samples
# -----------------------------
mean = images.mean(axis=0)
sd = images.std(axis=0, ddof=1)
eps = 1e-8
tvec = mean / np.maximum(sd / np.sqrt(n), eps)

# Run BOTH one-sided directions (recommended)
p_pos = np.clip(stats.t.sf(tvec, df=n - 1), 1e-300, 1.0)   # faces > houses
p_neg = np.clip(stats.t.cdf(tvec, df=n - 1), 1e-300, 1.0)  # faces < houses

# Choose which direction to proceed with:
# If you "intended" faces > houses, use p_pos; otherwise swap.
pmap = p_pos
direction_name = "pos (faces > houses)"

print("Direction:", direction_name)
print("min p:", float(pmap.min()), "mean p:", float(pmap.mean()))

# -----------------------------
# 5) Multiple comparisons: Bonf, BH, BY, BKY
# -----------------------------
alpha = 0.05
q_fdr = 0.05

# Bonferroni
bonf = pmap <= (alpha / m)

# Sort once
p_sorted = np.sort(pmap)

# BH
bh_line = (np.arange(1, m + 1) / m) * q_fdr
bh_ok = p_sorted <= bh_line
pcrit_bh = float(p_sorted[np.max(np.where(bh_ok)[0])]) if np.any(bh_ok) else np.nan
bh = (pmap <= pcrit_bh) if np.isfinite(pcrit_bh) else np.zeros_like(pmap, bool)

# BY
c_m = np.sum(1.0 / np.arange(1, m + 1))
by_line = (np.arange(1, m + 1) / m) * (q_fdr / c_m)
by_ok = p_sorted <= by_line
pcrit_by = float(p_sorted[np.max(np.where(by_ok)[0])]) if np.any(by_ok) else np.nan
by = (pmap <= pcrit_by) if np.isfinite(pcrit_by) else np.zeros_like(pmap, bool)

# BKY (two-stage)
q0 = q_fdr / (1.0 + q_fdr)
bky1_line = (np.arange(1, m + 1) / m) * q0
bky1_ok = p_sorted <= bky1_line
pcrit1 = float(p_sorted[np.max(np.where(bky1_ok)[0])]) if np.any(bky1_ok) else np.nan
R1 = int(np.sum(pmap <= pcrit1)) if np.isfinite(pcrit1) else 0
m0_hat = max(1, m - R1)
q_star = min(1.0, q0 * (m / m0_hat))

bky2_line = (np.arange(1, m + 1) / m) * q_star
bky2_ok = p_sorted <= bky2_line
pcrit_bky = float(p_sorted[np.max(np.where(bky2_ok)[0])]) if np.any(bky2_ok) else np.nan
bky = (pmap <= pcrit_bky) if np.isfinite(pcrit_bky) else np.zeros_like(pmap, bool)

print("\nRejections:")
print(" Bonf:", int(bonf.sum()))
print(" BH  :", int(bh.sum()), "pcrit:", pcrit_bh)
print(" BY  :", int(by.sum()), "pcrit:", pcrit_by)
print(" BKY :", int(bky.sum()), "pcrit:", pcrit_bky)

# -----------------------------
# 6) Cluster permutation (sign-flip across pseudo-samples)
# -----------------------------
p0_cluster = 0.01
n_perm = 1000
t0 = float(stats.t.isf(p0_cluster, df=n - 1))

mask_data = mask_img.get_fdata().astype(bool)
struct3 = np.ones((3, 3, 3), dtype=int)

def max_cluster_extent_3d(tvals_vec):
    t_img = unmask(tvals_vec, mask_img).get_fdata()
    supra = (t_img >= t0) & mask_data
    lab, nlab = ndimage.label(supra, structure=struct3)
    if nlab == 0:
        return 0, lab
    sizes = np.bincount(lab.ravel())[1:]
    return int(sizes.max(initial=0)), lab

# observed clusters
t_obs_vec = tvec
_, _ = max_cluster_extent_3d(t_obs_vec)

# null distribution of maximum cluster size
rng = np.random.default_rng(0)
null_max = np.empty(n_perm, dtype=int)

for b in range(n_perm):
    signs = rng.choice([-1.0, 1.0], size=(n, 1))
    imgp = images * signs
    meanp = imgp.mean(axis=0)
    sdp = imgp.std(axis=0, ddof=1)
    tp = meanp / np.maximum(sdp / np.sqrt(n), eps)
    null_max[b], _ = max_cluster_extent_3d(tp)

# threshold observed clusters by clusterwise p
t_img_obs = unmask(t_obs_vec, mask_img).get_fdata()
supra_obs = (t_img_obs >= t0) & mask_data
lab, nlab = ndimage.label(supra_obs, structure=struct3)

sig_cluster = np.zeros(mask_data.shape, dtype=bool)
if nlab > 0:
    sizes = np.bincount(lab.ravel())[1:].astype(int)
    for k, s in enumerate(sizes, start=1):
        p_cl = (1.0 + np.sum(null_max >= s)) / (n_perm + 1.0)
        if p_cl <= alpha:
            sig_cluster[lab == k] = True

print("\nClusterPerm:")
print(" clusters found:", int(nlab))
print(" sig vox:", int(sig_cluster.sum()))

# -----------------------------
# 7) Build NIfTI images for plotting
# -----------------------------
t_img = unmask(tvec, mask_img)
bh_img = unmask(bh.astype(int), mask_img)
clus_img = new_img_like(t_img, sig_cluster.astype(int))

# -----------------------------
# 8) Plot
# -----------------------------
plotting.plot_stat_map(
    t_img,
    bg_img=anat_img,
    threshold=2.0,
    title=f"t-map ({direction_name}), threshold t>=2"
)

plotting.plot_roi(
    bh_img,
    bg_img=anat_img,
    title="BH significant voxels (q=0.05)"
)

plotting.plot_roi(
    clus_img,
    bg_img=anat_img,
    title=f"ClusterPerm significant clusters (alpha=0.05, p0={p0_cluster})"
)

plotting.show()
```
